### print() Fonksiyonunun Parametreleri: sep, end, file, flush. Yıldızlı Parametreler

`print()` fonksiyonu, çeşitli parametreler alabilir. Bu parametreler, fonksiyonun davranışını değiştirebilir ve çıktının formatlanmasını sağlar.


1. **sep**: Bu parametre, çıktıda değerler arasında kullanılacak ayırıcıyı (seperator) belirler. Varsayılan olarak bir boşluktur (`' '`).

````python
print("Python", "Java", "C++", sep=", ")
````


2. **end**: Bu parametre, çıktının sonunda kullanılacak karakteri belirler. Varsayılan olarak bir satır sonudur (`'\n'`).


````python
print("Satırın sonu", end=" burada.")
````

Bu kod parçası, "Satırın sonu burada." metnini yazdırır ve bir sonraki `print` fonksiyonu aynı satırın devamına yazdırır.



3. **file**: Bu parametre, çıktının yazılacağı dosyayı belirler.


`file` parametresi, Python'daki `print()` fonksiyonunun çıktısını bir dosyaya yönlendirmek için kullanılır. Normalde, `print()` fonksiyonunun çıktısı standart çıktı akışı olan terminal veya komut istemcisine gönderilir. Yani varsayılan olarak `sys.stdout`'tur, yani çıktı standart çıktı akışına (etkileşimli kabuk (interactive shell) veya komut satırı (command window)) yönlendirilir. Ancak, `file` parametresi sayesinde, bu çıktıyı bir dosyaya veya başka bir çıktı akışına yönlendirebilirsiniz.

````python
# 'dosya.txt' isimli bir dosya açıyoruz. 'w' modu, dosyanın yazma modunda açılmasını sağlar.
# Eğer dosya mevcut değilse oluşturulur, mevcutsa içeriği silinir ve yeniden yazılır.

dosya = open("deneme.txt", "w")
print("Dosyaya yazılan metin", file=dosya)
dosya.close()

# with open('dosya.txt', 'w') as f:
#   print("Dosyaya yazılan metin", file=f)
````

- `file` parametresi olarak geçirilen dosya nesnesinin yazılabilir modda (`'w'`, `'a'`, `'x'` vb.) açılmış olması gerekir.


4. **flush**: Bu parametre, `True` olarak ayarlandığında, çıktının tamponlanmadan hemen diske yazılmasını sağlar. Varsayılan değeri `False`'tur.


````python
print("Hemen yazdır", flush=True)
````

````python
f = open("yeni_dosya.txt", "w")
````

````python
print("Ubuntu", file=f)
````

````python
print("Python", file=f)
````

print ile Ubuntu ve Python string'lerini dosyaya yazdırdık. Ancak şimdi yeni_dosya.txt adlı dosyayı açıp kontrol ettiğinizde içinin boş olduğunu göreceğiz. Çünkü bizim print ile eklediğimiz satırları Python tampona kaydetti. Dosyaya yazma işlemleri sona erdiğinde ise yani dosya kapatıldığında, Python tamponda bekleyen bütün bilgileri standart çıktı konumuna (yani bizim durumumuzda f adlı değişkenin tuttuğu yeni_dosya.txt adlı dosyaya) boşaltacak.

````python
f.close()
````

Görülebileceği üzere, Python öncelikle tüm yazılacak verileri bir tamponda tutar ve dosya kapatıldığında bu tampondaki verileri dosyaya aktarır. Flush parametresi, bu aktarma işleminin kontrolünü sağlar. Büyük verileri `print()` fonksiyonu aracılığıyla yazdırıp bunların derhal diske kaydedilmesini istiyorsanız, `flush=True` parametresini kullanmanız mümkündür.

Eğer bu parametreye bir değer atamazsak, Python varsayılan olarak bu değeri False olarak alacak ve verilerin dosyaya yazılabilmesi için dosyanın kapatılmasını bekleyecektir. ### Print içinde `*` opreatörünün unpacking işlevi.

````python
range(10)
````

````python
print(range(10))
````

````python
print(*range(10))
````

`print(range(10))` ifadesi bir range nesnesini yazdırırken, `print(*range(10))` ifadesi range nesnesi içindeki tüm sayıları unpack edip yazdırır. Bu, Python'da `*` operatörünün unpacking işlevinin bir örneğidir.

````python
enumerate("Python")
````

````python
print(*enumerate("Python"))
````

Bu ifade, `enumerate("Python")` tarafından oluşturulan enumerate nesnesini unpack eder ve bu nesnenin her bir elemanını ayrı ayrı argümanlar olarak `print` fonksiyonuna geçirir. `*` operatörü, nesnenin içindeki her bir çifti (yani index ve value çiftlerini) ayırır. Ancak, Python'da `enumerate` nesnesi doğrudan unpack edildiğinde, her bir çift ayrı bir argüman olarak kabul edilir ve `print` fonksiyonu, bu çiftleri boşlukla ayrılmış değerler olarak yazdırır. Bu nedenle, `print(*enumerate("Python"))` ifadesinin çıktısı beklenenin aksine biraz karışık olabilir; çünkü her index-value çifti parantezler ve virgüller olmadan yan yana yazdırılır ve bu, genellikle okunması zor bir çıktı üretir.

Daha anlaşılır bir çıktı için, enumerate nesnesi genellikle bir döngüde kullanılır.

````python
reversed("Python")
````

````python
print(*reversed("Python"))
````

## complex() Fonksiyonu

Python'da `complex()` fonksiyonu, karmaşık sayılar oluşturmak için kullanılır. Karmaşık sayılar, gerçek ve sanal bölümlerden oluşur ve genellikle `a + bj` formatında gösterilir, burada `a` gerçek kısmı, `b` sanal kısmı temsil eder ve `j` sanal birimin simgesidir (matematikte genellikle `i` olarak bilinir).

`complex()` fonksiyonu iki parametre alabilir: `real` ve `imag`. `real` gerçek kısmı, `imag` ise sanal kısmı belirtir. Her iki parametre de isteğe bağlıdır ve varsayılan değerleri sıfırdır.

````python
# Gerçek ve sanal kısımları olmayan bir karmaşık sayı
sayi1 = complex()
print(sayi1)
````

````python
# Gerçek kısmı olan ve sanal kısmı olmayan bir karmaşık sayı
sayi2 = complex(3)
print(sayi2)
````

````python
# Hem gerçek hem de sanal kısımları olan bir karmaşık sayı
sayi3 = complex(4, 5)
print(sayi3)
````

- Python'da karmaşık sayıların gerçek ve sanal kısımlarına erişmek için `.real` ve `.imag` öznitelikleri kullanılır.


````python
sayi = complex(3, 4)
print(sayi.real)
print(sayi.imag)
````

## eval() ve exec() Fonksiyonları

Python'da `eval()` ve `exec()` fonksiyonları, dinamik olarak Python ifadelerini ve kodlarını çalıştırmak için kullanılır. `eval()` ve `exec()` fonksiyonlarının isimleri, sırasıyla "evaluate" (değerlendirme) ve "execute" (yürütme) kelimelerinden türetilmiştir.



#### eval() Fonksiyonu:

`eval()` fonksiyonu, bir string olarak verilen bir Python ifadesini değerlendirir ve ifadenin sonucunu döndürür. Genellikle matematiksel hesaplamalar ve kullanıcı girdilerinden elde edilen değerlerin değerlendirilmesi gibi basit durumlar için kullanılır.

````python
x = 5
ifade = 'x * 5 + 10'
sonuc = eval(ifade)
print(sonuc)
````


Bu örnekte, `eval()` fonksiyonu `ifade` değişkenini değerlendirir ve `x * 5 + 10` hesaplamasının sonucunu döndürür.



#### exec() Fonksiyonu:
`exec()` fonksiyonu, `eval()`'a benzer şekilde, bir string içinde verilen Python kodunu çalıştırır. Ancak, `exec()` herhangi bir değer döndürmez (her zaman `None` döndürür). `exec()`, birden fazla Python ifadesini içeren kod bloklarının yürütülmesi için kullanılır. `eval()` ile gerçekleştiremeyeceğimiz bazı işlemleri `exec()` ile yapabiliriz. Özellikle, `exec()` fonksiyonu, karakter dizileri (stringler) içinde verilen daha geniş ve kompleks Python kodlarını çalıştırmamıza olanak tanır.

Örneğin, `eval()` kullanarak bir değişken tanımlaması yapamazsınız. Bu, `eval()`'ın yalnızca ifadeleri değerlendirmek için tasarlanmış olmasından kaynaklanır. Yani, `eval("x = 3")` şeklinde bir kod çalıştırdığınızda hata alırsınız. Ancak, `exec()` ile bu tür bir atama işlemi gerçekleştirebilirsiniz:



````python
exec("x = 3")
````

````python
print(x)
````

Bu kodu çalıştırdığınızda, `x` adında bir değişken tanımlanır ve bu değişkene 3 değeri atanır. `print(x)` dediğimizde ekrana 3 değeri yazdırılır. Bu, `exec()`'in değişken tanımlama gibi işlemleri de desteklediğinin bir göstergesidir.

````python
kod = """
for i in range(5):
    print(i)
"""
````

````python
exec(kod)
````

## Bool (Boolean) işleçleri
Bilgisayar biliminde temel bir kavramdır ve her şeyin iki temel değerden birine sahip olduğu bir sistem üzerine kuruludur: Doğru (True) veya Yanlış (False). Bu değerler, İngiliz matematikçi George Boole'un çalışmalarından esinlenerek adlandırılmıştır. Bu iki değer, programlama içinde koşulların ve ifadelerin doğruluğunu test etmek için kullanılır.

Örneğin, bir değişkenin değeri üzerinde durum sorgulaması yapalım:


````python
a = 10
print(a == 10)  # Bu ifade True (Doğru) döndürür çünkü a'nın değeri gerçekten de 10'dur.
print(a == 5)   # Bu ifade False (Yanlış) döndürür çünkü a'nın değeri 5 değil, 10'dur.
````

Python'da her ifadenin bir bool değeri vardır. Genel bir kural olarak, 0 ve boş değerler (`""`, `[]`, `{}`, `None` gibi) False olarak değerlendirilir. Bunun dışındaki tüm değerler True olarak kabul edilir. Bu durum, bool işleçlerinin programlamada çok yönlü kullanımını sağlar.

Bu konsepti daha iyi anlamak için bool() fonksiyonu ile birkaç örnek yapalım:


````python
print(bool(1))
````

````python
print(bool(-42))     # Negatif sayılar da dahil olmak üzere sıfırdan farklı tüm sayılar için, çıktı: True
````

````python
print(bool(0))
````

````python
bool("0")   # 0 string'i için, çıktı: True
````

````python
print(bool("Python")) # Boş olmayan herhangi bir string için, çıktı: True
````

````python
print(bool(""))      # Boş bir string için, çıktı: False
````



Bu örneklerden görebileceğiniz gibi, `bool()` fonksiyonu bir değerin doğruluğunu veya yanlışlığını kontrol etmek için kullanılabilir. Bu, koşullu ifadelerde, döngülerde ve diğer kontrol yapılarında oldukça yararlıdır.

````python
ad = input("Adınızı giriniz: ")

if bool(ad): # bool yazmasak'da olur
    print("Adınız başarıyla kaydedildi!")
else:
    print("Ad alanı boş bırakılamaz!")
````

## Walrus operatörü, `:=`

Python 3.8 ile tanıtılan Walrus operatörü, `:=` işaretleriyle temsil edilir. Bu yeni atama ifadesi, bir değişkene değer atarken aynı zamanda bu değeri bir ifade içinde kullanmanıza olanak tanır. Bu, kodunuzda hem yerden tasarruf etmenizi hem de okunabilirliği artırmanızı sağlar.

Walrus operatörünü kullanmadan önce, bir değeri hesaplamanız ve bu değeri bir değişkene atamanız, sonra da bu değişkeni bir ifade içinde kullanmanız gerekirdi. Walrus operatörü sayesinde, bu iki adımı tek bir adımda gerçekleştirebilirsiniz.

````python
yas = int(input("Yaşınızı girin: "))
if yas < 18:
    print("Üzgünüm, bu etkinliğe katılmak için yaşınızın 18 veya üzeri olması gerekiyor.")
else:
    print("Hoş geldiniz! Etkinliğe katılmaya uygunsunuz.")
````

````python
if (yas := int(input("Yaşınızı girin: "))) < 18:
    print("Üzgünüm, bu etkinliğe katılmak için yaşınızın 18 veya üzeri olması gerekiyor.")
else:
    print("Hoş geldiniz! Etkinliğe katılmaya uygunsunuz.")
````

### Aitlik İşleçleri (Membership Operators) `in` ve `not in`

Python'da aitlik işleçleri, bir öğenin bir dizi, liste, demet ya da benzeri bir koleksiyonun içinde olup olmadığını kontrol etmek için kullanılır. En yaygın aitlik işleçleri `in` ve `not in`dır.


````python
harfler = ['a', 'b', 'c', 'd']

# 'in' kullanımı
if 'a' in harfler:
    print("'a' harfler listesinin bir üyesidir.")

# 'not in' kullanımı
if 'z' not in harfler:
    print("'z' harfler listesinin bir üyesi değildir.")
````

Bu örneklerde, `'a'` harfinin `harfler` listesinin bir elemanı olup olmadığı kontrol edilir. İkinci durumda ise, `'z'` harfinin `harfler` listesinde olmadığının kontrolü yapılır.


### Kimlik İşleçleri (Identity Operators) `is` ve `is not`

Kimlik işleçleri, iki nesnenin bellekte aynı nesne olup olmadığını kontrol etmek için kullanılır.

Python'da kimlik işleçleri `is` (... midir?) ve `is not` (... değil midir?) şeklindedir.


Python'da her nesnenin, bellekteki benzersiz adresini temsil eden bir kimlik numarası vardır. Bu kimlik numarasına `id()` fonksiyonu ile ulaşabiliriz. `id()` fonksiyonu, nesnenin bellek adresini temsil eden benzersiz bir tam sayı döndürür.

````python
a = 42
print(id(a))  # 'a' nesnesinin kimlik numarası
````

````python
b = [1, 2, 3]
print(id(b))  # 'b' nesnesinin kimlik numarası
````

````python
c = "Python"
print(id(c))  # 'c' nesnesinin kimlik numarası
````

````python
# 'a' ve 'b' değişkenlerinin farklı nesneler olması
d = 42
print(id(d))  # 'd' değişkeni de 'a' gibi 42 değerine sahip ancak kimlikleri farklı olabilir
````

````python
# Listelerde durum
e = b
print(id(e))  # 'e' ve 'b' aynı nesneyi referans alır, dolayısıyla kimlik numaraları aynı olacaktır.
````


Bu örneklerde, `id()` fonksiyonu kullanılarak farklı nesnelerin kimlik numaralarına ulaşılır. İki değişken aynı nesneyi referans aldığında (örneğin `e` ve `b`), aynı kimlik numarasına sahip olurlar. Ancak, iki değişkenin içeriği aynı olsa bile (örneğin `a` ve `d`), eğer bunlar ayrı nesnelerse farklı kimlik numaralarına sahip olacaklardır.

````python
a = [1, 2, 3]
b = a
c = [1, 2, 3]

print("a'nın id:",id(a))

print("b'nin id:",id(b))

print("c'nin id:",id(c))
````

````python
# 'is' kullanımı
if b is a: # b, a mıdır?
    print("b ve a aynı nesneyi referans alıyor.")

# 'is not' kullanımı
if c is not a: # c, a değil midir?
    print("c ve a farklı nesneleri referans alıyor.")
````

Bu örneklerde, `b` değişkeni `a` değişkenine eşitlendiğinden, her ikisi de aynı liste nesnesini referans alır ve `b is a` ifadesi `True` (Doğru) döner. Ancak, `c` değişkeni `a` ile aynı değerlere sahip yeni bir liste oluşturduğu için, `c is not a` ifadesi `True` (Doğru) döner çünkü her iki değişken de bellekte farklı nesneleri referans alır.

Bu işleçler, özellikle nesnelerin aynı nesne olup olmadığını kontrol etmek veya bir öğenin bir koleksiyonun parçası olup olmadığını belirlemek için kullanılır ve Python programlamada sıkça karşılaşılan durumlardır.

## Python'da `_` (alt çizgi) ifadesi

Python'da, `_` karakteri genellikle atanan değeri kullanmayacağımızı belirtmek için kullanılır. Bu, özellikle çoklu değer döndüren fonksiyonlarla çalışırken veya döngülerde bazı değerleri görmezden gelmek istediğimizde yararlı olabilir.


### 1. Çoklu Değer Atamada `_` Kullanımı
Bir fonksiyon birden fazla değer döndürdüğünde ve bu değerlerden sadece birkaçını kullanmak istediğimizde, geri kalanı için `_` kullanabiliriz.

````python
def get_user_data():
    # Kullanıcı adı, yaş ve cinsiyet döndürür
    return "Ahmet", 30, "Erkek"

isim, _, _ = get_user_data()  # Sadece ismi kullanıyoruz, diğer bilgiler önemli değil
print(isim)
````

### 2. Döngülerde `_` Kullanımı
Bir döngüde, iterasyon değişkeni olarak `_` kullanarak, bu değişkenin önemli olmadığını ve kullanılmayacağını belirtebiliriz.


````python
for _ in range(5):
    print("Merhaba!")  # 5 kez "Merhaba!" yazdırır, döngüdeki sayı önemli değil
````


### 3. Tuple'da Belirli Değerleri Atamada `_` Kullanımı
Bir tuple içindeki belirli değerleri atarken, ilgilenmediğimiz değerler için `_` kullanabiliriz.


````python
bilgiler = ("Ahmet", "Yılmaz", 35, "Mühendis")  # İsim, soyisim, yaş, meslek
isim, soyisim, _, _ = bilgiler  # Sadece isim ve soyisimle ilgileniyoruz
print(f"{isim} {soyisim}")
````


### 4. Listelerde `_` Kullanımı
Listelerde belirli bir indexe sahip elemanları atarken, `_` kullanabiliriz. Ancak, bu daha az yaygındır çünkü genellikle bu tür durumlar için liste dilimleme kullanılır. Ancak, özellikle paketlenmiş değerlerle çalışırken kullanılabilir.



````python
_, x, *_ = [1, 2, 3, 4, 5]  # Sadece ikinci elemanla ilgileniyoruz
print(x)
````

Bu örnekler, Python'da `_` kullanımının çeşitli yönlerini göstermektedir. Genellikle, bu kullanım okunabilirliği artırır ve ilgilenmediğimiz değerleri açıkça gösterir.

### Enumerate Fonksiyonu:

Python'da `enumerate()` fonksiyonu, bir dizi, liste, demet ya da metin gibi yinelenebilir (iterable) nesneler üzerinde döngü yaparken, öğelerin hem kendisini hem de sıra numarasını (index) elde etmemizi sağlar. İngilizcede "enumerate" kelimesi "numaralandırmak" anlamına gelir. Bu fonksiyon, her bir öğe için bir çift (tuple) döndürür: İlk öğe sıra numarası, ikinci öğe ise yinelenebilir nesnenin öğesidir.

### `enumerate()` Fonksiyonunun Temel Kullanımı:
`enumerate()` fonksiyonunun genel yapısı şu şekildedir:

```python
enumerate(yinelenebilir, start=0)
```

Bu yapıda, `yinelenebilir`, bir liste, demet, dize veya başka bir yinelenebilir yapı olabilir. Opsiyonel olarak `start` parametresi, saymaya başlanacak indeks numarasını belirler; eğer belirtilmezse varsayılan olarak 0'dan başlar.

### `enumerate()` Fonksiyonu ile İlgili Örnekler:

**Liste Elemanlarının ve İndekslerinin Yazdırılması**:

````python
liste = ["Elma", "Armut", "Muz", "Kiraz"]
for indeks, eleman in enumerate(liste):
    print(indeks, eleman)
````

Bu örnekte, `liste`nin her bir elemanı ve onun indeksi, `enumerate()` fonksiyonu kullanılarak yazdırılır. Sonuç olarak, her bir meyve ve onun sırası konsola basılır.

**Bir stringin indeksi ile birlikte yazılması:**

````python
for index, value in enumerate("Python"):
    print(index, value)
````

Bu örnekte, `enumerate()` fonksiyonu "Python" kelimesinin her bir harfini ve her harfin sıra numarasını döndürür. Döngü her çalıştığında, `index` değişkeni harfin sıra numarasını, `value` değişkeni ise harfi kendisini alır.

````python
print(*enumerate("Python"))
````

````python
list(enumerate("Python"))
````

**Çift İndekslere Sahip Elemanların Yazdırılması**:

````python
liste = ["a", "b", "c", "d", "e", "f", "g"]
for indeks, eleman in enumerate(liste):
    if indeks % 2 == 0:
        print("Eleman:", eleman)
````

Bu örnekte, `liste`deki çift indeksli elemanlar (indeks 0, 2, 4, ...) `enumerate()` fonksiyonu ile filtrelendirilerek yazdırılır. Bu sayede, sadece "a", "c", "e", "g" gibi çift indeksli elemanlar ekrana basılır.

### Sonuç:
`enumerate()` fonksiyonu, for döngüleri içinde hem elemanlara hem de onların indekslerine erişmek istediğimizde oldukça faydalıdır. Bu fonksiyon, kodumuzun okunabilirliğini ve anlaşılırlığını artırır, aynı zamanda manuel olarak indeks yönetimi yapmamızı gerektiren karmaşık yapıları ortadan kaldırır. Özellikle veri işleme, veri analizi ve düzenli veri yapılarının işlenmesi sırasında `enumerate()` fonksiyonundan yararlanabilirsiniz. Bu fonksiyonun sağladığı esneklik ve kolaylık, Python programlamada önemli bir avantaj sağlar.

### Karakter Dizilerini Ters Çevirmek:

Karakter dizilerini tersine çevirmek için Python'da dilimleme (`slicing`) yöntemini kullanabilirsiniz:


````python
sehir = "istanbul"
print(sehir[::-1])
````

Bu kod, "istanbul" kelimesini tersine çevirerek "lubnatsi" olarak yazdırır. Burada `[::-1]` dilimleme işlemi, stringin başından sonuna kadar tersten okunmasını sağlar.


**Reversed() Fonksiyonu:**

`reversed()` fonksiyonu, yinelenebilir bir nesnenin elemanlarını tersten sıralar. Ancak, bu fonksiyon doğrudan bir liste veya string döndürmez, bunun yerine bir iterator döndürür. Bu iterator'ün içeriğini görmek için onu bir döngü içinde kullanabilir veya bir liste ya da string'e dönüştürebilirsiniz.


````python
for harf in reversed("Merhaba Python"):
    print(harf, end="")
````

Bu örnek, "Merhaba Python" ifadesini ters çevirerek yazdırır. `reversed()` fonksiyonu, ifadenin her bir karakterini tersten sıralar ve `for` döngüsü bu karakterleri yazdırır.

Alternatif olarak, `print()` fonksiyonunu ve `*` operatörünü kullanarak ters çevrilmiş karakterleri aynı satırda yazdırabilirsiniz:

````python
print(*reversed("Merhaba Python"), sep="")
````

Bu kod da "Merhaba Python" ifadesinin ters çevrilmiş halini yazdırır. `sep=""` argümanı, karakterler arasında herhangi bir ayrım olmadan bunların yan yana yazdırılmasını sağlar.

## Python'da `chr()` fonksiyonu

Bir tam sayının (ASCII kodu) karşılık geldiği karakteri döndürmek için kullanılır. Bu fonksiyon, 0'dan 1,114,111'e (0x10FFFF) kadar olan tam sayı değerlerini alabilir ve bu tam sayıların Unicode karakter karşılıklarını döndürür. Temel olarak, `chr()` fonksiyonu tam sayıyı karaktere çevirir.

````python
# ASCII kodu 97 olan karakteri bulalım (ki bu 'a' harfidir).
harf = chr(97)
print(harf)
````

````python
# ASCII kodu 65 olan karakteri bulalım (ki bu 'A' harfidir).
baska_bir_harf = chr(65)
print(baska_bir_harf)
````

`chr()` fonksiyonu özellikle, belirli bir ASCII veya Unicode kod aralığına karşılık gelen karakterleri elde etmek istediğinizde kullanışlıdır. Bu, özellikle metin işleme, şifreleme, veri formatlama veya özel karakterlerin kullanılması gereken durumlarda faydalıdır.

- `chr()` fonksiyonunun tam tersi `ord()` fonksiyonudur. `ord()` fonksiyonu bir karakter alır ve o karakterin ASCII (veya Unicode) tam sayı karşılığını döndürür.
- Eğer `chr()` fonksiyonuna 0'dan küçük veya 1,114,111'den büyük bir değer verirseniz, bir `ValueError` hatası alırsınız, çünkü bu değerler geçerli Unicode karakter aralığının dışındadır.

## Python'da `ord()` fonksiyon

Python'da `ord()` fonksiyonu, verilen tek bir karakterin Unicode (veya ASCII) karşılığını tam sayı olarak döndürmek için kullanılır. Bu fonksiyon, bir karakterin tam sayı değerini elde etmek istediğinizde yararlıdır.

````python
# 'a' karakterinin Unicode (ASCII) tam sayı değerini bulalım
deger = ord('a')
print(deger)
````

````python
# 'A' karakterinin Unicode (ASCII) tam sayı değerini bulalım
baska_deger = ord('A')
print(baska_deger)
````

````python
# Türkçe karakter örneği
turkce_karakter_degeri = ord('ş')
print(turkce_karakter_degeri)  # Türkçe 'ş' karakterinin Unicode tam sayı değerini yazdırır
````

Bu örneklerde, `ord()` fonksiyonu çeşitli karakterlerin Unicode değerlerini döndürür. İlk iki örnekte, İngiliz alfabesinin küçük 'a' ve büyük 'A' harflerinin ASCII değerleri sırasıyla 97 ve 65 olarak döndürülür. Üçüncü örnekte ise, Türkçe bir karakter olan 'ş'nin Unicode değeri döndürülür.

- `ord()` fonksiyonu yalnızca tek karakterlik stringler ile çalışır. Eğer stringin uzunluğu birden fazla karakter ise `TypeError` hatası alırsınız.
- `ord()` fonksiyonunun tersi `chr()` fonksiyonudur. `chr()` tam sayıları karakterlere çevirirken, `ord()` karakterleri tam sayılara çevirir.
