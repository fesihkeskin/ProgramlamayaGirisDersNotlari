**Soru:**
Bir derste dersi alan her öğrenci için iki sınav sonucu (0-100 arası) ve bir proje notu (0-100 arası) verildiğini varsayınız. Sınav notlarının %30'u ve proje notunun %40'ı öğrencinin son notunu oluşturur. Aşağıda sırasıyla öğrencilerin adı, sınav 1 sonucu, sınav 2 sonucu ve proje notunu içeren bir liste verilmiştir:

```python
ogrenciler = [
    ("Ali", 75, 80, 85),
    ("Ayşe", 90, 85, 88),
    ("Mehmet", 65, 70, 75),
    ("Elif", 80, 75, 82),
    ("Ahmet", 95, 90, 92)
]
```
Bu öğrencilerin son notlarını hesaplayarak, her bir öğrencinin adı ve son notunu içeren bir sözlük oluşturun ve ekrana yazdırın. Daha sonra son notlarına göre en yüksek notu alan öğrenciyi bulun ve ekrana yazdırın.

**İpucu:**
Her öğrenci için son notu hesaplamak için aşağıdaki formülü kullanabilirsiniz:

```python
son_not = (sınav1 * 0.3) + (sınav2 * 0.3) + (proje_notu * 0.4)
```

Sonra her bir öğrencinin adını ve son notunu bir sözlük içinde saklayınız. Son olarak, bu sözlüğü kullanarak en yüksek notu alan öğrenciyi bulabilirsiniz.

***Beklenen ekran çıktısı:***
```
Ali: 80.50
Ayşe: 87.70
Mehmet: 70.50
Elif: 79.30
Ahmet: 92.30
En yüksek notu alan öğrenci: Ahmet (92.30 puan)
```

````python
ogrenciler = [
    ("Ali", 75, 80, 85),
    ("Ayşe", 90, 85, 88),
    ("Mehmet", 65, 70, 75),
    ("Elif", 80, 75, 82),
    ("Ahmet", 95, 90, 92)
] # 5 adet tuple içeren bir listedir.

sonuclar = {}  # Öğrenci adlarını ve sonuçlarını saklayacak boş bir sözlük oluşturun

for ogrenci in ogrenciler: # her tuple'ı tek tek ele alalım
    ad = ogrenci[0] # ele alınan tuple'ın ilk elemanını ad değişkenine at
    sınav1 = ogrenci[1]
    sınav2 = ogrenci[2]
    proje_notu = ogrenci[3]

    son_not = (sınav1 * 0.3) + (sınav2 * 0.3) + (proje_notu * 0.4)

    sonuclar[ad] = son_not # Öğrenci adı (key) ve son notunu (value) sözlüğe ekleyin

# Öğrencilerin sonuçlarını ekrana yazdırın
for ogrenci, son_not in sonuclar.items():
    print(f"{ogrenci}: {son_not:.2f}")

# En yüksek notu alan öğrenciyi bulun
en_yuksek_notlu_ogrenci = max(sonuclar, key=sonuclar.get)
en_yuksek_not = sonuclar[en_yuksek_notlu_ogrenci]

print(f"En yüksek notu alan öğrenci: {en_yuksek_notlu_ogrenci} ({en_yuksek_not:.2f} puan)")
````

**Soru:**
Bir restoran menüsü oluşturmak istiyorsunuz. Kullanıcıdan menüde bulunacak yemeklerin isimlerini ve fiyatlarını girmesini isteyen bir program yazın. Kullanıcı her yemek ekledikten sonra başka bir yemek eklemek isteyip istemediğini sorun. Eğer kullanıcı eklemek istemiyorsa, menüyü ekrana yazdırın.

***Örnek olarak, iki yemek ve fiyatı ekledikten sonra "Beklenen ekran çıktısı:"***
```
Yemek ismi: şakşuka
Fiyatı (TL): 5
Başka bir yemek eklemek ister misiniz? (Evet/Hayır): evet
Yemek ismi: iskender
Fiyatı (TL): 15
Başka bir yemek eklemek ister misiniz? (Evet/Hayır): hayır

Menü:
şakşuka: 5.0 TL
iskender: 15.0 TL
```


````python
# Boş bir menü sözlüğü oluşturun
menu = {}

# Kullanıcıdan yemek ve fiyat girişi alın
while True:
    yemek = input("Yemek ismi: ")
    fiyat = float(input("Fiyatı (TL): "))

    # Menüye yemeği ve fiyatını ekleyin
    menu[yemek] = fiyat

    devam = input("Başka bir yemek eklemek ister misiniz? (Evet/Hayır): ")
    if devam.lower() != "evet":
        break

# Menüyü ekrana yazdırın
print("\nMenü:") # \n (newline yani yeni satıra (alt satır) geç demek)
for yemek, fiyat in menu.items():
    print(f"{yemek}: {fiyat} TL")
````

**Soru:** `kelime_sozlugu_olustur()`isimli bir fonksiyon yazacaksınız. Fonksiyon ile bir kullanıcıdan birden fazla kelime alarak bir sözlük oluşturacaksınız. Her kelimenin ilk harfini anahtar olarak kullanın ve değer olarak o harfle başlayan tüm kelimelerin listesini saklayın. Eğer bir anahtar için birden fazla kelime varsa, bu kelimeleri listeye ekleyin. Kullanıcı 'çık' kelimesini girdiğinde, oluşturulan sözlüğü ve her anahtar için kaç kelime olduğunu gösteren bir raporu yazdırın ve programı sonlandırın.

***Örnek olarak, kullanıcı aşağıdaki üç kelimeyi giriyor sonra çık diyerek bitiriyor bu durumda "Beklenen ekran çıktısı:"***
```
Bir kelime girin veya 'çık' yazarak bitirin: gelmek
Bir kelime girin veya 'çık' yazarak bitirin: gitmek
Bir kelime girin veya 'çık' yazarak bitirin: uçmak
Bir kelime girin veya 'çık' yazarak bitirin: çık
'g' harfiyle başlayan 2 kelime: ['gelmek', 'gitmek']
'u' harfiyle başlayan 1 kelime: ['uçmak']
Sözlüğümüz: {'g': ['gelmek', 'gitmek'], 'u': ['uçmak']}
```

````python
def kelime_sozlugu_olustur():
    sozluk = {}  # Anahtar: İlk harf, Değer: Kelime listesi

    # Kullanıcının sözlüğe kaç kelime gireceğini bilmediğimiz için sonsuz
    # bir while döngüsü başlatırız. sadece 'çık' yazılırsa döngü kırılacak

    while True:

        # Kullanıcıdan kelimeyi al ve strip metodu ile kelime'in başındaki
        # ve sonundaki boşlukları kaldır.
        kelime = input("Bir kelime girin veya 'çık' yazarak bitirin: ").strip()

        if kelime == 'çık':  # Çıkış koşulu gerçekleşirse while'ı kır
            break

        ilk_harf = kelime[0].lower()  # İlk harfi al ve küçük harfe çevir

        # Sözlüğe ilk harf (key) ve kelimeyi (value) ekle
        if ilk_harf in sozluk: # anahtar olarak ilk_harf sözlükte zaten var ise,
            sozluk[ilk_harf].append(kelime)
        else: # anahtar olarak ilk_harf sözlükte yok ise,
            sozluk[ilk_harf] = [kelime]

    # Raporu yazdır
    for harf, kelimeler in sozluk.items():
        print(f"'{harf}' harfiyle başlayan {len(kelimeler)} kelime: {kelimeler}")
    return print("Sözlüğümüz:",sozluk)

# Fonksiyonu çağır
kelime_sozlugu_olustur()
````

**Soru:**

Bir derste, öğrencilerin başarı durumları 'Dönem Sonu Notu'na göre belirlenir. Her öğrenci için, vize ve final sınavlarından alınan notlar 100 tam puan üzerinden değerlendirilir. Öğrencinin Dönem Sonu Notu, vize sınav notunun %40'ı ve final sınav notunun %60'ının toplamı alınarak hesaplanır. Dersi başarıyla tamamlamak için, Final sınav notu en az 50 olmalı ve Dönem Sonu Notunun en az 60 olması  gerekmektedir. Yani eğer bir öğrencinin final notu 50'den küçükse dönem sonu notuna bakılmaksızın durumunu doğrudan "Başarısız" olarak belirleyeceğiz. Aşağıda, dönem sonu notu aralığına göre alınacak harf notları ve öğrencinin durumu verilmiştir:

**Harf Notu (Durumu)   : Dönem Sonu Notu**
- AA      (Mükemmel) : 90-100
- BA      (Çok İyi)  : 80-89
- BB      (İyi)      : 70-79
- CB      (Orta)     : 65-69
- CC      (Başarılı) : 60-64
- FF      (Başarısız): 0-59

Aşağıda, bazı öğrencilerin isimleri, vize ve final sınav notları yer almaktadır:

```
ogrenciler = [
    ("Ali", 75, 80),
    ("Ayşe", 90, 85),
    ("Mehmet", 65, 70),
    ("Elif", 80, 75),
    ("Ahmet", 95, 90)
]
```

Bu verileri kullanarak, her bir öğrenci için Dönem sonu notunu, harf notunu ve notun durumunu içeren bir **sözlük** oluşturun. Bu sözlüğü ekrana yazdırın. Daha sonra, bu öğrenciler arasında en yüksek dönem sonu notuna sahip olan öğrenciyi belirleyin ve ekrana yazdırın.

````python
ogrenciler = [ ("Ali", 75, 80), ("Fatma", 95, 43), ("Ayşe", 90, 85), ("Mehmet", 65, 70), ("Elif", 30, 65), ("Ahmet", 95, 90) ]

# Öğrencilerin dönem sonu notları hesaplama, harf notları ve durumunu belirleme
sonuclar = {}
for ogrenci in ogrenciler:
    ad, vize, final = ogrenci
    donem_sonu_notu = vize * 0.4 + final * 0.6

    # Eğer final notu 50'den küçükse, doğrudan başarısız
    if final < 50:
        harf, durum = "FF", "Başarısız"
    else:
        # Harf notu ve durum belirleme
        if 90 <= donem_sonu_notu <= 100:
            harf, durum = "AA", "Mükemmel"
        elif 80 <= donem_sonu_notu < 90:
            harf, durum = "BA", "Çok İyi"
        elif 70 <= donem_sonu_notu < 80:
            harf, durum = "BB", "İyi"
        elif 65 <= donem_sonu_notu < 70:
            harf, durum = "CB", "Orta"
        elif 60 <= donem_sonu_notu < 65:
            harf, durum = "CC", "Başarılı"
        else:  # 0-59 aralığı
            harf, durum = "FF", "Başarısız"

    sonuclar[ad] = {"Dönem Sonu Notu": donem_sonu_notu, "Harf": harf, "Durum": durum}
# En yüksek notu alan öğrenciyi bulma
en_yuksek_notlu_ogrenci = None
max_not = -1
for ad, bilgiler in sonuclar.items():
    if bilgiler["Dönem Sonu Notu"] > max_not:
        max_not = bilgiler["Dönem Sonu Notu"]
        en_yuksek_notlu_ogrenci = (ad, bilgiler)
print("En yüksek notlu öğrenci:",en_yuksek_notlu_ogrenci[0],", Notu:",en_yuksek_notlu_ogrenci[1]['Dönem Sonu Notu'])
````

**Soru:**

Verilen bir `tuple`(demet) içindeki elemanlardan birden fazla kez tekrar edenleri belirleyen bul_ve_say(demetim) fonksiyonu yazınız. Programınız, `tuple` içindeki her bir elemanı incelemeli ve eğer bir eleman birden fazla kez varsa, bu elemanın adını ve kaç kez tekrar ettiğini içeren bir sözlük oluşturmalıdır.

Örneğin, verilen `tuple` şu şekilde olsun: `('a', 'b', 'c', 'd', 'e', 'c', 'c', 'a', 'a', 'd', 'd', 'd')`. Programınızın sonucunda oluşturulması gereken sözlük şöyle olmalıdır: `{'a': 3, 'c': 3, 'd': 4}`. Bu sözlük, her bir elemanın `tuple` içinde kaç kez tekrar ettiğini göstermektedir.

````python
def bul_ve_say(demetim):
    sayac = {}
    i = 0
    while i < len(demetim):
        eleman = demetim[i]
        if demetim.count(eleman) > 1:
            sayac[eleman] = demetim.count(eleman)
        i += 1
    return sayac

# Örnek tuple
ornek_tuple = ('a', 'b', 'c', 'd', 'e', 'c', 'c', 'a', 'a', 'd', 'd', 'd')

# Sonucu hesapla ve yazdır
sonuc = bul_ve_say(ornek_tuple)
print(sonuc)
````

````python
def bul_ve_say(demetim):
    tekrar_sayaci = {}
    for eleman in demetim:
        if demetim.count(eleman) > 1:
            tekrar_sayaci[eleman] = demetim.count(eleman)

    # Sadece birden fazla kez tekrar eden elemanları içeren öğeleri döndür
    return {k: v for k, v in tekrar_sayaci.items() if v > 1}

# Örnek tuple ve fonksiyonun kullanımı
ornek_tuple = ('a', 'b', 'c', 'd', 'e', 'c', 'c', 'a', 'a', 'd', 'd', 'd')
sonuc = bul_ve_say(ornek_tuple)
print(sonuc)

````
