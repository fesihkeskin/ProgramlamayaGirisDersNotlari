**Soru 1: Öğrenci Not Takip Uygulaması**

Bir üniversite için öğrencilerin vizelerini, finallerini kaydedip, harf notlarını hesaplayan ve yöneten bir program yazmanız istenmektedir.Bu işlemleri gerçekleştirmek için dört ana fonksiyon yazmanız gerekmektedir:

1. **not_oku()** - 'notgirisi.txt' dosyasından öğrenci notlarını okur. Eğer dosya boşsa, "Herhangi bir not bulunmamaktadır" mesajını yazdırır.
2. **not_gir()** - Kullanıcıdan öğrencinin adını, soyadını, vize ve final notlarını alır ve bu bilgileri 'notgirisi.txt' dosyasına ekler. Bu bilgileri `"isim soyisim: vize, final\n"` şeklinde  bir formatta kaydetmenizde fayda var. Bu format, verilerin düzenli bir şekilde saklanmasını ve sonradan okunmasını kolaylaştırır.
3. **not_kaydet()** - Okunan notlar üzerinden her öğrenci için harf notunu hesaplar ve bu sonuçları 'harf_notlari.txt' dosyasına kaydeder.
4. **not_sil()** - Kullanıcıya tüm notları silmek isteyip istemediği sorulur ve onaylanırsa, ilgili not dosyaları temizlenir.
5. **tek_not_harf_hesapla()** - Kullanıcıdan tek bir vize ve final notu alınır ve bu notlara göre harf notu hesaplanır. Bu fonksiyon, özellikle bireysel öğrenciler için hızlı bir not değerlendirmesi yapılmasında kullanışlıdır.

Kullanıcıya, bu fonksiyonlardan hangisini kullanmak istediğini soran bir menü sunulur ve çıkmak için 'q' seçeneği verilir. Uygulama, kullanıcı çıkmak isteyinceye kadar döngü içinde çalışır.


**Menü içeriği:**
```Python
1-Notları Oku
2-Yeni Not Gir
3-Harf Notu Hesapla ve Kaydet
4-Notları Sil
5-Tek Öğrenci İçin Harf Notu Hesapla
-Çıkış için 'q' yaz:
```

````python
import time

def not_oku():
    with open("notgirisi.txt","r",encoding="utf-8") as file:
        satirlar = file.readlines()
        if not satirlar:
            print('Herhangi bir not bulunmamaktadır')
        else:
            return satirlar

def not_gir():
    isim = input('İsim giriniz:')
    soyisim = input('Soyisim giriniz:')
    vize = input('Vize Notunu giriniz:')
    final = input('Final Notunu giriniz:')
    with open("notgirisi.txt","a",encoding="utf-8") as file:
        satir = f"{isim} {soyisim}:{vize},{final}\n"
        file.write(satir)

def not_kaydet():
    satirlar = not_oku()
    sonuclar = []
    if satirlar is not None:
        for satir in satirlar:
            sonuclar.append(harf_hesapla(satir))
    with open("harf_notlari.txt","w",encoding="utf-8") as file:
        file.writelines(sonuclar)
    return sonuclar

def not_sil():
    kontrol = input("Tüm notları silmek istediğinize emin misiniz? (e/h):")
    if kontrol.lower() == "e":
        with open("notgirisi.txt","w",encoding="utf-8") as file:
            pass
        with open("harf_notlari.txt","w",encoding="utf-8") as file:
            pass

def harf_hesapla(satir):
    satir = satir.strip()
    isim_soyisim, notlar = satir.split(':')
    vize, final = map(int, notlar.split(','))
    ort = vize * 0.4 + final * 0.6
    if final < 50:
        harf, durum = "FF", "Başarısız"
    elif ort >= 90:
        harf, durum = "AA", "Mükemmel"
    elif ort >= 80:
        harf, durum = "BA", "Çok İyi"
    elif ort >= 70:
        harf, durum = "BB", "İyi"
    elif ort >= 65:
        harf, durum = "CB", "Orta"
    elif ort >= 60:
        harf, durum = "CC", "Başarılı"
    else:
        harf, durum = "FF", "Başarısız"
    return f"{isim_soyisim}: {ort}, {harf}, {durum}\n"

def tek_harf_notu_hesapla():
    tek_not = input("Vize ve final notu giriniz (örneğin 80,90): ")
    satir = ':'+ tek_not
    print(harf_hesapla(satir))

while True:
    islem = input("\n1-Notları Oku\n2-Yeni Not Gir\n3-Harf Notu Hesapla ve Kaydet\n4-Notları Sil\n5-Tek Öğrenci İçin Harf Notu Hesapla\n-Çıkış için 'q' yaz:\n")
    if islem == "1":
        satirlar = not_oku()
        print("-"*6 + 'Notlar' + "-"*6)
        if satirlar:
            print(''.join(satirlar))
        print("-"*20)
    elif islem == "2":
        not_gir()
    elif islem == "3":
        sonuclar = not_kaydet()
        print("-"*6 + 'Harf Notları' + "-"*6)
        print(''.join(sonuclar))
        print("-"*30)
    elif islem == "4":
        not_sil()
    elif islem == "5":
        tek_harf_notu_hesapla()
    elif islem == "q":
        print("*"*15)
        print("* Çıkılıyor.. *")
        print("*"*15)
        time.sleep(1)
        break
    else:
        print("1-5 arası bir seçenek seçiniz")
````

**Soru 2: Öğrenci Not Takip Uygulamasını Geliştirme**

Daha önce oluşturduğumuz Öğrenci Not Takip Uygulamasını geliştirerek, öğrencilerin başarı durumlarına göre "kalanlar.txt" ve "geçenler.txt" adında iki ayrı dosyaya bilgilerini yazınız. Öğrencilerin harf notları hesaplandıktan sonra, başarılı olan öğrenciler "geçenler.txt" dosyasına, başarısız olanlar ise "kalanlar.txt" dosyasına kaydedilecek.

**Geliştirme Adımları:**

1. **Harf Notu Hesapla ve Kaydet** fonksiyonunu genişleterek, öğrencilerin başarı durumlarına göre ayrıştırılması.
2. Başarılı ve başarısız öğrencilerin notlarını sırasıyla "geçenler.txt" ve "kalanlar.txt" dosyalarına yazdırma.

````python
import time

def not_kaydet_ve_ayristir():
    satirlar = not_oku()
    gecenler = []
    kalanlar = []
    if satirlar is not None:
        for satir in satirlar:
            sonuc, harf = harf_hesapla_ve_ayristir(satir)
            if harf == "FF":
                kalanlar.append(sonuc)
            else:
                gecenler.append(sonuc)

    with open("harf_notlari.txt", "w", encoding="utf-8") as file:
        file.writelines(satirlar)

    with open("gecenler.txt", "w", encoding="utf-8") as file:
        file.writelines(gecenler)

    with open("kalanlar.txt", "w", encoding="utf-8") as file:
        file.writelines(kalanlar)

def harf_hesapla_ve_ayristir(satir):
    satir = satir.strip()
    isim_soyisim, notlar = satir.split(':')
    vize, final = map(int, notlar.split(','))
    ort = vize * 0.4 + final * 0.6
    if final < 50:
        harf, durum = "FF", "Başarısız"
    elif ort >= 90:
        harf, durum = "AA", "Mükemmel"
    elif ort >= 80:
        harf, durum = "BA", "Çok İyi"
    elif ort >= 70:
        harf, durum = "BB", "İyi"
    elif ort >= 65:
        harf, durum = "CB", "Orta"
    elif ort >= 60:
        harf, durum = "CC", "Başarılı"
    else:
        harf, durum = "FF", "Başarısız"
    return f"{isim_soyisim}: {ort}, {harf}, {durum}\n", harf

# Öğrencilerin harf notlarını hesapla ve geçen/kalanları kaydet
not_kaydet_ve_ayristir()
````

1. **not_kaydet_ve_ayristir()**:
   - Mevcut notlar okunur.
   - Her bir öğrenci için `harf_hesapla_ve_ayristir()` fonksiyonu çağrılır.
   - Öğrencinin başarı durumuna göre, sonuçlar uygun listelere (`gecenler` veya `kalanlar`) eklenir.
   - Son olarak, bu listeler ilgili dosyalara yazılır.

2. **harf_hesapla_ve_ayristir()**:
   - Her öğrencinin notu hesaplanır.
   - Harf notu ve başarı durumu belirlenir.
   - Başarı durumu ve harf notu bilgileri geri döndürülür.
