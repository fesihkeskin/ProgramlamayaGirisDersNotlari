### Bubble sort Sıralama algoritması:

**Bubble sort**, basit bir sıralama algoritmasıdır ve adını, liste içindeki elemanların sıvı içindeki kabarcıklar gibi yukarı doğru "kabarcıklaşması" (yani yer değiştirmesi) işleminden alır. Algoritma, liste boyunca tekrar tekrar geçer ve her seferinde yan yana bulunan elemanları karşılaştırır. Eğer bir eleman, yanındaki elemandan büyükse, bu iki eleman yer değiştirir. Bu işlem, liste boyunca tekrarlanarak elemanlar doğru sıraya ulaşana kadar devam eder.

Örneğin Liste elemanlarımız `[3, 1, 2]` olsun. Bubble sort algoritması, her adımda yan yana bulunan elemanları karşılaştırır ve gerektiğinde yer değiştirir. Bu işlem, liste sıralanana kadar tekrarlanır.

Başlangıç:  `[3, 1, 2]`

- İlk eleman (3) ile ikinci eleman (1) karşılaştırılır. 3 > 1 olduğu için yer değiştirirler: `[1, 3, 2]`
- İkinci eleman (3) ile üçüncü eleman (2) karşılaştırılır. 3 > 2 olduğu için yer değiştirirler: `[1, 2, 3]`
- Bu noktada, en büyük eleman en sağa taşınmış oldu. Liste boyunca bir geçiş tamamlandı ve liste zaten sıralı.

Bubble sort algoritması, en kötü senaryoda \(n-1\) geçiş yapar, burada \(n\) liste boyutudur. Ancak, liste zaten sıralı olduğunda, daha fazla işlem yapmaya gerek yoktur. 3 elemanlı bir liste için, algoritmanın en fazla 2 geçiş yapması gerekebilir, ancak bu örnekte ilk geçişten sonra liste sıralandı.

Bubble sort algoritmasının temel adımları şunlardır:

1. Listeyi baştan sona doğru gez.
2. Her adımda, şu anki elemanı ile bir sonraki elemanı karşılaştır.
3. Eğer şu anki eleman bir sonraki elemandan büyükse, yerlerini değiştir.
4. Tüm liste sıralanana kadar 1. adımdan itibaren işlemleri tekrarla.

**Soru 1: Bubble sort algoritması**

Girdi olarak bir liste alan ve bu listeyi bubble sort algoritması ile sıralayıp geri döndüren bir `bubble_sort(liste)` fonksiyonu yazınız. Fonksiyonunuzu test etmek için, `random.randint` fonksiyonunu kullanarak rastgele sayılardan oluşan bir listeyi **list comprehension ile `for` döngüsünü** kullanarak oluşturun ve bu listeyi fonksiyonunuza girdi olarak verin. Yazdığınız `bubble_sort(liste)` fonksiyonun, verilen listeyi başarıyla sıraladığını gözlemleyin.

**İpucu:** Python'da rastgele sayılar üretmek için `random.randint(a, b)` fonksiyonu kullanılır. Bu fonksiyon, `a` ve `b` arasında (her iki değer dahil) rastgele bir tamsayı döndürür. Örneğin, `random.randint(1, 100)` ifadesi, 1 ile 100 arasında rastgele bir tamsayı üretir.


**Beklenen Örnek Ekran Çıktısı:**

```
Orijinal Liste: [8, 1, 2, 3, 9, 1, 8, 7, 5, 10]
Sıralı Liste: [1, 1, 2, 3, 5, 7, 8, 8, 9, 10]
```

````python
import random

def bubble_sort(liste):
    n = len(liste)
    for i in range(n-1):
        for j in range(0, n-i-1): # Burada range(0, n-1)yazsak da olur, fakat neden n-i-1 yazdığımızı düşünün
            if liste[j] > liste[j+1]:
                liste[j], liste[j+1] = liste[j+1], liste[j] # eğer büyükse yer değiştir
    return liste

# Rastgele sayılardan oluşan bir liste oluştur
rastgele_liste = [random.randint(1, 10) for _ in range(10)]
print("Orijinal Liste:", rastgele_liste)

# Bubble sort fonksiyonunu kullanarak listeyi sırala
sirali_liste = bubble_sort(rastgele_liste)
print("Sıralı Liste:", sirali_liste)
````

**Soru 2: Fonksiyon Çalışma Süresini Hesaplama**

Yazılan bir fonksiyonun çalışma süresini hesaplayan ve sonucu konsola yazdıran bir `zaman_dekoratoru(fonksiyon)` dekoratörü yazınız. Dekore edilen fonksiyonların `__name__` ve `__doc__` gibi özelliklerini koruyunuz.Dekoratörünüzü, birinci soruda yazdığınız `bubble_sort(liste)` fonksiyonu ve aşağıda verilen Python'un built-in `sorted()` fonksiyonu üzerinde test ederek çalışma zamanlarını ekrana yazdırınız.

```Python
def timsort(liste):
    """ Python'un built-in sıralama fonksiyonu"""
    return sorted(liste)
```

**İpucu:** `time` modülünü kullanarak fonksiyonların başlangıç ve bitiş zamanını `time.time()` kaydedebilirsiniz. `functools.wraps` fonksiyonunu kullanarak, dekore edilen fonksiyonun `__name__` ve `__doc__` gibi özelliklerini koruyabilirsiniz.


**Beklenen Örnek Ekran Çıktısı:**

```
tim_sort fonksiyonunun çalışma süresi: 4.5299530029296875e-06 saniye.
[0, 1, 2, 4, 4, 5, 5, 6, 10, 10]
```
```
bubble_sort fonksiyonunun çalışma süresi: 1.6450881958007812e-05 saniye.
[0, 1, 2, 4, 4, 5, 5, 6, 10, 10]
```

````python
import time, random
from functools import wraps

def zaman_dekoratoru(fonksiyon):
    @wraps(fonksiyon)
    def wrapper(*args, **kwargs):
        """Wrapper fonksiyonu"""
        baslangic = time.time()
        sonuc = fonksiyon(*args, **kwargs)
        bitis = time.time()
        print(f"{fonksiyon.__name__} fonksiyonunun çalışma süresi: {bitis - baslangic} saniye.")
        return sonuc
    return wrapper

@zaman_dekoratoru
def tim_sort(liste):
    """Python'un built-in sıralama fonksiyonu"""
    return sorted(liste)

@zaman_dekoratoru
def bubble_sort(liste):
    """Bubble sort algoritması"""
    n = len(liste)
    for i in range(n-1):
        for j in range(0, n-i-1):
            if liste[j] > liste[j+1]:
                liste[j], liste[j+1] = liste[j+1], liste[j]
    return liste
````

````python
# Test etmek için bir liste oluşturalım
rastgele_liste = [random.randint(0, 10) for _ in range(10)]
rastgele_liste
````

````python
# tim_sort fonksiyonunu çağıralım
tim_sort(rastgele_liste)
````

````python
# tim_sort fonksiyonunun docstring'i korundu
print(tim_sort.__doc__)
````

````python
# bubble_sort fonksiyonunu çağıralım
bubble_sort(rastgele_liste)
````

````python
# bubble_sort fonksiyonunun docstring'i korundu
print(bubble_sort.__doc__)
````

**Soru 3: Fonksiyona Loglama Ekleme**

Bir fonksiyona giriş ve çıkış anlarında log mesajları yazdırmak için bir dekoratör yazınız. Bu dekoratör, fonksiyonun adını, aldığı argümanları ve döndürdüğü değeri loglamalıdır (kaydını tutmalıdır). Dekore edilen fonksiyonların `__name__` ve `__doc__` gibi özelliklerini koruyunuz.

Soru 2 de yazdığınız `@zaman_dekoratoru` ve şimdi yazdığınız `@loglama_dekoratoru`'nu, iki sayının toplamını ve çarpımını hesaplayan fonksiyonlar ile test ediniz.


**Beklenen Örnek Ekran Çıktısı:**

```
[2024-04-01 21:00:35.406557] 'toplama' fonksiyonu çağrıldı. Dönüş değeri: 7
toplama fonksiyonunun çalışma süresi: 0.001100301742553711 saniye.
7
```
```
[2024-04-01 21:00:35.432887] 'carpma' fonksiyonu çağrıldı. Dönüş değeri: 30
carpma fonksiyonunun çalışma süresi: 0.00011372566223144531 saniye.
30
```

````python
from datetime import datetime
from functools import wraps

def loglama_dekoratoru(fonksiyon):
    @wraps(fonksiyon) # Dekore edilen fonksiyonların __name__ ve __doc__ gibi özelliklerini korumak için
    def wrapper(*args, **kwargs):
        """Wrapper fonksiyonudur."""
        cagrilma_zamani = datetime.now() #  Greenwich başlangıç zamanıdırç
        sonuç = fonksiyon(*args, **kwargs)
        print(f"[{cagrilma_zamani}]: {fonksiyon.__name__} fonksiyonu çağrıldı. Dönüş değeri: {sonuç}")
        return sonuç
    return wrapper

@zaman_dekoratoru # bir önceki soruda yazdığımız zaman dekoratörünü çağıralım
@loglama_dekoratoru
def toplama(a, b):
    """İki sayının toplamını döndürür."""
    return a + b

@zaman_dekoratoru
@loglama_dekoratoru
def carpma(a, b):
    """İki sayının çarpımını döndürür."""
    return a * b
````

````python
# toplama fonksiyonunu çağıralım
toplama(3, 4)
````

````python
# toplama fonksiyonunun docstring'i
print(toplama.__doc__)
````

````python
# çarpma fonksiyonunu çağıralım
carpma(5, 6)
````

````python
# çarpma fonksiyonunun docstring'i
print(carpma.__doc__)
````

**Soru 4: Erişim Kontrolü Yapma**

Belirli bir kullanıcı listesine fonksiyon erişimi sağlayan bir dekoratör yazın. Dekoratör, yalnızca belirlenen kullanıcı listesinde bulunan kullanıcılar için fonksiyonun çalıştırılmasına izin vermeli, diğer kullanıcılarda ise bir erişim reddi mesajı döndürmelidir.

Dekoratörünüzü test etmek için aşağıdaki gibi bir gizli fonksiyon yazınız:
```Python
@erisim_kontrolu
def gizli_fonksiyon(kullanici, mesaj):
    return f"Kullanıcı: {kullanici}, Gizli mesaj: {mesaj}"
```
ve
```Python
gizli_fonksiyon('admin', 'Merhaba Dünya!')
```
şeklinde çağırınız.

**İpucu:** Dekoratörünüzde `izin_verilen_kullanicilar = ['admin', 'ahmet', 'ayşe']` şeklinde bir kullanıcı listesi tanımlayabilir ve fonksiyon çağrılırken geçirilen kullanıcının bu listede olup olmadığını wrapper iç fonksiyonda kontrol edebilirsiniz.

**Beklenen Örnek Ekran Çıktısı:**

```
Kullanıcı: admin, Gizli mesaj: Merhaba Dünya!
```
ya da
```
Üzgünüm, mehmet erişim iznine sahip değilsiniz.
```

````python
def erisim_kontrolu(fonksiyon):
    izin_verilen_kullanicilar = ['admin', 'ahmet', 'ayşe']
    def wrapper(*args, **kwargs):
        if args[0] in izin_verilen_kullanicilar:
            return fonksiyon(*args, **kwargs)
        else:
            return f"Üzgünüm, {args[0]} erişim iznine sahip değilsiniz."
    return wrapper

@erisim_kontrolu
def gizli_fonksiyon(kullanici, mesaj):
    return f"Kullanıcı: {kullanici}, Gizli mesaj: {mesaj}"
````

````python
# Fonksiyonu test et
print(gizli_fonksiyon('admin', 'Merhaba Dünya!'))
````

````python
gizli_fonksiyon('ahmet', 'Merhaba Dünya!')
````

````python
print(gizli_fonksiyon('mehmet', 'Merhaba Dünya!'))
````

**Ek not:**

**`datetime.now()` fonksiyonu**, Python'un `datetime` modülünde bulunur ve mevcut tarih ve saat bilgisini almak için kullanılır. Bu fonksiyon, yerel zaman dilimine göre mevcut tarih ve saat bilgilerini içeren bir `datetime` nesnesi döndürür. Kullanımı oldukça basittir ve çeşitli senaryolarda, özellikle uygulamaların zamanla ilgili işlemleri gerçekleştirmesi gerektiğinde kullanılır.

**Kullanımı:**

Öncelikle `datetime` modülünü içe aktarırız:

````python
from datetime import datetime
````

Ardından, mevcut tarih ve saat bilgisini almak için `datetime.now()` fonksiyonunu çağırabilirsiniz:

````python
su_an = datetime.now()
print("Şu anki tarih ve saat:", su_an)
````

Bu kod parçası, yerel zaman dilimine (Greenwich) göre mevcut tarih ve saat bilgisini ekrana yazdırır.

- Eğer UTC (Coordinated Universal Time, Koordinatlı Evrensel Zaman) cinsinden mevcut zamanı almak istiyorsanız, `datetime.utcnow()` fonksiyonunu kullanabilirsiniz. Bu, zaman dilimi farklılıklarını dikkate almadan evrensel bir zaman sağlar.
- `datetime.now()` fonksiyonu, opsiyonel olarak bir `tz` (timezone, zaman dilimi) parametresi alabilir. Bu parametre ile belirli bir zaman dilimine göre mevcut zamanı alabilirsiniz. Bunun için Python'un `pytz` (Python time zone) kütüphanesi gibi zaman dilimi bilgilerini içeren bir kütüphane kullanmanız gerekir.

````python
from datetime import datetime
import pytz

istanbul_zamani = datetime.now(pytz.timezone('Europe/Istanbul'))
print("İstanbul'daki şu anki tarih ve saat:", istanbul_zamani)
````

`datetime.now()` fonksiyonu, loglama, kullanıcı etkinliklerinin zaman damgalarını kaydetme, zaman tabanlı işlemlerin gerçekleştirilmesi gibi çeşitli durumlar için oldukça yararlıdır. Bu fonksiyon sayesinde, yazılımınızın zamanla ilgili hassas ve doğru işlemler yapması sağlanır.

UTC±00.00, hesaplamaları için referans noktası İngiltere, Greenwich'teki 0° başlangıç meridyeni üzerindeki zaman dilimi referans alınır. GMT (Greenwich Mean Time, Greenwich Ortalama Zamanı) ile aynı anlamda kullanılır.
