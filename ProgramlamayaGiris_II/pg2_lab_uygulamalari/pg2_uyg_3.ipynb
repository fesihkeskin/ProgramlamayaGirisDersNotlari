**Soru 1:**

`numbers` adında bir tam sayılar dizisi ve `target` adında bir hedef tam sayı verildiğinde, toplamları hedef sayıya eşit olan iki sayının dizideki konumlarını (indislerini) bulacak bir `two_sum(numbers, target)` adında bir fonksiyon yazınız. Her veri seti için tam olarak bir çözüm olduğunu varsayabilirsiniz ve bir eleman yalnızca bir kez kullanılabilir. Sonuçları herhangi bir sırayla döndürebilirsiniz.

Bu fonksiyonu, aşağıdaki örneklere göre test ediniz:

**Örnekler:**

1. **Girdi:** numbers = [2, 7, 11, 15], target = 9
   **Çıktı:** [0, 1]
   **Açıklama:** `numbers[0]` + `numbers[1]` (2 + 7) toplamı 9'a eşittir, bu nedenle [0, 1] çiftini döndürüyoruz.

2. **Girdi:** numbers = [3, 2, 4], target = 6
   **Çıktı:** [1, 2]
   **Açıklama:** `numbers[1]` + `numbers[2]` (2 + 4) toplamı 6'ya eşittir, bu nedenle [1, 2] çiftini döndürüyoruz.

3. **Girdi:** numbers = [3, 3], target = 6
   **Çıktı:** [0, 1]
   **Açıklama:** `numbers[0]` + `numbers[1]` (3 + 3) toplamı 6'ya eşittir, bu nedenle [0, 1] çiftini döndürüyoruz.

Lütfen `two_sum(numbers, target)` fonksiyonunun doğru çıktıyı ürettiğinden emin olmak için bu girişleri kullanarak fonksiyonunuzu test edin. Fonksiyon, belirtilen hedefe ulaşmak için gereken iki sayının indekslerini bir liste olarak döndürmelidir.

```python
from typing import List

def two_sum(numbers: List[int], target: int) -> List[int]:
    """
    Verilen bir tam sayılar dizisi içerisinde,
    toplamları belirli bir hedef değere eşit olan iki farklı sayının indekslerini bulur.

    Args:
    numbers: Tam sayılardan oluşan bir dizi.
    target: Hedeflenen toplam değeri.

    Returns
    İki sayının indekslerinden oluşan bir liste. Eğer uygun iki sayı bulunamazsa, boş bir liste döndürülür.
    """
    length = len(numbers)  # Dizinin uzunluğunu alır.
    for i in range(length - 1):  # Birinci sayı için döngü.
        for j in range(i + 1, length):  # İkinci sayı için döngü.
            if numbers[i] + numbers[j] == target:  # Toplamı kontrol eder.
                return [i, j]  # Uygun ilk iki indeksi döndürür.
    return []  # Eğer uygun bir çift bulunamazsa, boş bir liste döndürür.
```

```python
# Test örneklerini kullanarak fonksiyonu çağırıyoruz.
print(two_sum([2, 7, 11, 15], 9))
print(two_sum([3, 2, 4], 6))
print(two_sum([3, 3], 6))
```

**Soru 2:** Bir tam sayı (`x`) verildiğinde, bu sayının palindrom olup olmadığını kontrol eden `isPalindrome(x)` fonskiyonu yazın. Bir sayı soldan sağa ve sağdan sola doğru aynı şekilde okunuyorsa, bu sayı bir palindromdur.

**Örnekler:**
- **Girdi:** x = 121
  **Çıktı:** True
  **Açıklama:** 121, her iki yönden de aynı şekilde okunur.

- **Girdi:** x = -121
  **Çıktı:** False
  **Açıklama:** -121 soldan sağa okunduğunda farklıdır; sağdan sola 121- şeklindedir.

- **Girdi:** x = 10
  **Çıktı:** False
  **Açıklama:** 10, sağdan sola 01 olarak okunur; bu yüzden bir palindrom değildir.

```python
def isPalindrome(x: int) -> bool:
    if x < 0:
        return False
    temp, original = 0, x
    while x > 0:
        temp = temp * 10 + x % 10
        x //= 10
    return temp == original
```

```python
isPalindrome(121)
```

**Soru 3:** Bir merdiveni tırmanıyorsunuz ve zirveye ulaşmak için `n` basamak çıkmanız gerekiyor. Her adımda bir veya iki basamak çıkabilirsiniz. Basamak sayısını girdiğinizde Merdivenin zirvesine kaç farklı şekilde ulaşabileceğinizi döndüren `climbStairs(n)` fonskiyonu yazın?

**Örnekler:**
- **Girdi:** n = 0
  **Çıktı:** 1
  **Açıklama:** Zirveye ulaşmak için hiç basamak tırmanmanıza gerek yoktur, dolayısıyla sadece bir yol vardır: hiç çıkmamak.

- **Girdi:** n = 1
  **Çıktı:** 1
  **Açıklama:** Zirveye ulaşmak için sadece bir basamak tırmanmanız gerekiyorsa, yalnızca bir yol vardır: tek bir adımda tırmanmak.

- **Girdi:** n = 2
  **Çıktı:** 2
  **Açıklama:** İki farklı tırmanış yolu vardır: (1 adım + 1 adım) veya (2 adım).

- **Girdi:** n = 3
  **Çıktı:** 3
  **Açıklama:** Üç farklı tırmanış yolu vardır: (1 adım + 1 adım + 1 adım), (1 adım + 2 adım) veya (2 adım + 1 adım).

```python
def climbStairs(n: int) -> int:
    if n == 0 or n == 1:
        return 1
    prev, curr = 1, 1
    for i in range(2, n + 1):
        temp = curr
        curr = prev + curr
        prev = temp
    return curr
```

```python
climbStairs(3)
```

Bu fonksiyonda kullanılan `prev`, `curr` ve `temp` değişkenleri, merdiveni tırmanma problemi için kullanılan dinamik programlama yaklaşımında geçici değerleri saklamak için kullanılmaktadır. İsimlendirme, bu değişkenlerin rollerini ve kullanıldıkları bağlamı açıklamak amacıyla seçilmiştir.

- `prev`(previous = önceki): Bu değişken, bir önceki adımda ulaşılan farklı yolların sayısını temsil eder.
- `curr` (current = şimdiki): Bu değişken, şu anki adımda ulaşılan farklı yolların sayısını temsil eder.
- `temp` (temporary = geçici): Bu değişken, bir sonraki adımda kullanılmak üzere geçici olarak `curr` değerini saklar.

Örnek bir adım adım açıklama için 3 basamaklı bir merdiveni tırmanma durumunu ele alalım:

1. `prev` ve `curr` başlangıçta her ikisi de `1` olarak başlatılır.
2. `i = 2` için:
   - `temp = curr = 1`: Geçici değişken olan `temp`, şu anki adımda ulaşılan farklı yolların sayısını (`curr`) saklar.
   - `curr = prev + curr = 1 + 1 = 2`: Şu anki adımda ulaşılan farklı yolların sayısı, bir önceki adımda ulaşılan farklı yolların sayısı (`prev`) ile şu anki adımda ulaşılan farklı yolların sayısının (`curr`) toplamıdır.
   - `prev = temp = 1`: Bir sonraki adımda kullanılacak olan önceki adımın farklı yollarının sayısını temsil eden `prev`, geçici değişken `temp` ile güncellenir.
3. `i = 3` için:
   - `temp = curr = 2`: Geçici değişken olan `temp`, şu anki adımda ulaşılan farklı yolların sayısını (`curr`) saklar.
   - `curr = prev + curr = 1 + 2`: Şu anki adımda ulaşılan farklı yolların sayısı, bir önceki adımda ulaşılan farklı yolların sayısı (`prev`) ile şu anki adımda ulaşılan farklı yolların sayısının (`curr`) toplamıdır.
   - `prev = temp = 2`: Bir sonraki adımda kullanılacak olan önceki adımın farklı yollarının sayısını temsil eden `prev`, geçici değişken `temp` ile güncellenir.

Bu adımlar, her bir döngü tekrarı için `prev`, `curr` ve `temp` değişkenlerinin nasıl güncellendiğini göstermektedir. Bu sayede, her adımda merdiveni tırmanmanın farklı yollarını hesaplayabilir ve istenen `n` basamağına ulaşmanın farklı yollarının sayısını bulabiliriz.

**Not:** Bu yaklaşım Fibonacci dizisinde kullanılır. Fibonacci dizisi, her bir elemanın kendisinden önceki iki elemanın toplamı olduğu bir sayı dizisidir. Başlangıçta 0 ve 1 ile başlar ve sırasıyla 0, 1, 1, 2, 3, 5, 8, 13, 21, ... şeklinde devam eder.

Örneğin, Fibonacci dizisinin ilk birkaç elemanını düşünelim:
- 0. eleman: 0
- 1. eleman: 1
- 2. eleman: 0 + 1 = 1
- 3. eleman: 1 + 1 = 2
- 4. eleman: 1 + 2 = 3
- 5. eleman: 2 + 3 = 5
- ve böyle devam eder.

Görüldüğü gibi, her eleman kendisinden önceki iki elemanın toplamıdır. İşte bu özelliği nedeniyle Fibonacci dizisi, merdiveni tırmanma problemiyle (veya adımları birleştirme problemiyle) aynı matematiksel özelliklere sahiptir.

Merdiveni tırmanma problemi gibi, Fibonacci dizisi de dinamik programlama kullanarak etkili bir şekilde çözülebilir. Her adımda, şu anki eleman ve bir önceki eleman saklanır ve bir sonraki elemanın hesaplanmasında kullanılır. Bu da, merdiveni tırmanma problemi ile Fibonacci dizisi arasında benzerlik gösterir. Dolayısıyla, bu yaklaşım her ikisi için de geçerlidir.

```python
# for döngüsündeki iteratörü döngü içinde kullanmadığımız için
# burada i yerine _ yazabiliriz.
# Tabi burada önemli olan döngünün kaç defa çalışacağıdır.
# bu problemde döngü n-1 defa çalışmaktadır
def climbStairs(n: int) -> int:
    if n == 0 or n == 1:
        return 1
    prev, curr = 1, 1
    for _ in range(n-1):
        temp = curr
        curr = prev + curr
        prev = temp
    return curr
```

```python
climbStairs(3) # aynı sonucu verecektir
```

**Ek bilgi:** Bu örnekte `for` döngüsündeki `i` döngü değişkeni (iteratörü, indeksi) döngü içinde kullanılmadığı için yerine `_`kullanılmıştır. Sonuçta bizim için burada önemli olan döngünün n-1 defa çalışmasıdır.

```python
# Döngünün n-1 defa çalıştığını anlamak için
# aşağıdaki iki for döngüsünü karşılaştırın
# her iki döngünün de n-1 defa çalıştığını göreceksiniz
n=5

print("Birinci for döngüsü")
for i in range(2, n + 1):
  print(i)

print("\n\nİkinci for döngüsü")
for i in range(n-1):
  print(i)

print("\n\n_ ile for döngüsü")
i=0
for _ in range(n-1):
  i += 1
  print(i)

print(f"""\nGördüğünüz üzere her üç durumda da for döngüsü {n-1} defa çalıştı
Bu örnekte istediğimiz döngüyü kullanabilirdik.
Sonuçta döngü iterasyonu ile işimiz yok,
bizim için önemli olan for döngüsünün kaç defa çalışacağıdır.""")
```
