**Soru 1:**

Bir sayıyı başka bir sayı ile çarpma işlemini, toplama işlemleri kullanarak gerçekleştiren bir Python programını `while` ve sonra `for` ile yazınız. Bu program, kullanıcıdan iki tam sayı almalı: birinci çarpan, ikinci çarpan olacak. Ardından, aşağıdaki adımları takip ederek çarpma işlemini gerçekleştiriniz:

1. Kullanıcıdan Birinci Çarpan ve ikinci çarpan sayıyı alarak, Birinci çarpan sayıyı, ikinci çarpan sayı kadar kez toplayın.
2. Toplama işlemini, ikinci çarpan sayının değeri kadar kez tekrarlayın.
3. Yaptığınız toplama işlemlerinin sonucu, çarpma işleminin sonucudur.

Örneğin: 3 ile 4'ü çarpmak için ardışık toplama işlemleri aşağıdaki gibi yapılır:
1. 3 + 3 + 3 + 3 = 12
Sonuç olarak, çarpım = 12 olacaktır.

***Beklenen Ekran Çıktısı:***
```
Birinci çarpan sayıyı giriniz: 3
İkinci çarpan sayıyı giriniz: 4
Çarpım: 12
```

**Cevap 1:**

````python
# Kullanıcıdan birinci ve ikinci çarpan sayı değerlerini al
# while çözümü
carpan1 = int(input("Birinci çarpan sayıyı giriniz: "))
carpan2 = int(input("İkinci çarpan sayıyı giriniz: "))

# Çarpma işlemi için değişkenler
sonuc = 0
sayac = 0

# Çarpma işlemini toplama kullanarak yap (while döngüsü ile)
while sayac < carpan2:
    sonuc += carpan1
    sayac += 1

# Sonucu yazdır
print(f"Çarpım: {sonuc}")
````

````python
# Kullanıcıdan birinci ve ikinci çarpan sayı değerlerini al
# for çözümü
carpan1 = int(input("Birinci çarpan sayıyı giriniz: "))
carpan2 = int(input("İkinci çarpan sayıyı giriniz: "))

# Çarpma işlemini toplama kullanarak yap (for döngüsü ile)
sonuc = 0
for _ in range(carpan2):
    sonuc += carpan1

# Sonucu yazdır
print(f"Çarpım: {sonuc}")
````

**Soru 2:**

Bir sayıyı başka bir sayıya bölme işlemini, çıkarma işlemleri kullanarak gerçekleştiren bir Python programını `while` ve sonra `for` ile yazınız. Bu program, kullanıcıdan iki tam sayı almalı: birinci sayı bölünen, ikinci sayı bölen olacak. Ardından, aşağıdaki adımları takip ederek bölme işlemini gerçekleştiriniz:

1. Kullanıcıdan Bölünen sayı ve bölen sayıyı alarak Bölünen sayıdan bölen sayıyı çıkarın.
2. Çıkarma işlemini, kalan bölünen sayı bölen sayıdan küçük veya ona eşit olana kadar tekrarlayın.
3. Yaptığınız çıkarma işlemi sayısını not edin; bu, bölme işleminin sonucudur (bölüm).
4. Çıkarma işlemleri sonrası kalan sayı, kalan olarak kaydedilecektir.

Örneğin: 5'i 2'ye bölmek için ardışık çıkarma işlemleri aşağıdaki gibi yapılır:
1. 5 - 2 = 3 (Bir kez çıkarma)
2. 3 - 2 = 1 (İki kez çıkarma)
Sonuç olarak, bölüm = 2 ve kalan = 1 olacaktır.

***Beklenen Ekran Çıktısı:***
```
Bölünen sayıyı giriniz: 5
Bölen sayıyı giriniz: 2
Bölüm: 2
Kalan: 1
```

**Cevap 2:**

````python
# Kullanıcıdan bölünen ve bölen değerlerini al
bölünen = int(input("Bölünen sayıyı giriniz: "))
bölen = int(input("Bölen sayıyı giriniz: "))

# Bölme işlemini çıkarma kullanarak yap
bölüm = 0  # Bölme işleminin sonucu
kalan = bölünen  # Her çıkarma işleminden sonra kalan miktar

# Bölünen, bölen'den büyük veya eşit olduğu sürece döngüyü çalıştır
while kalan >= bölen:
    kalan -= bölen  # Kalanı güncelle
    bölüm += 1  # Bölüm sayısını artır

# Sonuçları yazdıralım
print(f"Bölüm: {bölüm}, Kalan: {kalan}")
````

````python
bölünen = int(input("Bölünen sayıyı giriniz: "))
bölen = int(input("Bölen sayıyı giriniz: "))

# Çıkarma işlemlerinin maksimum sayısını hesaplayalım
maks_çıkarma_sayısı = bölünen

# Bölme işlemini gerçekleştirelim
bölüm = 0  # Bölme işleminin sonucu
kalan = bölünen  # Başlangıçta kalan, bölünenin kendisidir

# For döngüsü kullanarak çıkarma işlemlerini gerçekleştirelim
for _ in range(maks_çıkarma_sayısı):
    if kalan >= bölen:
        kalan -= bölen  # Kalanı güncelle
        bölüm += 1  # Bölüm sayısını artır
    else:
        break  # Eğer kalan, bölen'den küçükse döngüyü sonlandır

# Sonuçları yazdıralım
print(f"Bölüm: {bölüm}, Kalan: {kalan}")
````

**Soru 3:**

Pozitif tam sayılar olan X ve Y için, eğer X sayısının çarpanları (kendisi hariç) toplamı Y sayısına ve aynı zamanda Y sayısının çarpanları (kendisi hariç) toplamı X sayısına eşitse, bu sayılar dost sayılardır. Girilen iki sayının dost olup olmadığını kontrol eden bir Python programını yazınız.

***Yönerge:***
1. 'carpanlar_toplami' isimli bir fonksiyon yazarak, verilen bir sayının kendisi hariç pozitif çarpanlarının toplamını bulunuz.
2. Kullanıcıdan iki pozitif tam sayı alınız.
3. Her bir sayının çarpanlarını bulunuz ve 'carpanlar_toplami' fonksiyonunu kullanarak kendisi hariç çarpanlarının toplamını hesaplayınız.
4. Eğer X sayısının çarpanları toplamı Y sayısına ve Y sayısının çarpanları toplamı X sayısına eşitse, bu sayıların dost olduğunu belirtiniz.
5. Aksi takdirde, sayıların dost olmadığını belirtiniz.

***Beklenen Ekran Çıktısı:***
```
X sayısını giriniz: 220
Y sayısını giriniz: 284
220 ve 284 dost sayılardır.
```

Ya da

```
X sayısını giriniz: 20
Y sayısını giriniz: 30
20 ve 30 dost sayılar değildir.
```

**Cevap 3:**

````python
def carpanlar_toplami(sayi):
    toplam = 0
    for i in range(1, sayi):
        if sayi % i == 0:
            toplam += i
    return toplam

# Kullanıcıdan iki sayı al
x = int(input("X sayısını giriniz: "))
y = int(input("Y sayısını giriniz: "))

# Çarpanlar toplamını hesapla
x_toplam = carpanlar_toplami(x)
y_toplam = carpanlar_toplami(y)

# Dost sayı kontrolü
if x_toplam == y and y_toplam == x:
    print(f"{x} ve {y} dost sayılardır.")
else:
    print(f"{x} ve {y} dost sayılar değildir.")
````
