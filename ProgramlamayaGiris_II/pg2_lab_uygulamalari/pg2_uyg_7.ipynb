**Soru 1:**
Bir metni ters çeviren ve girdi metin tipinde değilse Python'nun standart `ValueError` hatasının mesajını "Lütfen metin tipinde bir veri girin." şeklinde güncelleyerek hata fırlatan bir fonksiyon yazınız.

Fonksiyonu kullanırken, oluşabilecek istisnaları yakalayacak ve kullanıcıya uygun bir hata mesajı gösterecek şekilde bir istisna yönetimi (exception handling) yapısı kurunuz. Yani fonksiyonu çağırırken çalışma zamanını durduracak olan `ValueError` hatasını `try`-`except` yapısı ile ele alınız.

````python
def ters_cevir(s):
    if not isinstance(s, str):
        raise ValueError("Lütfen metin tipinde bir veri girin.")
    return s[::-1]
````

````python
ters_cevir(15)
````

````python
ters_cevir('Python')
````

````python
# Fonksiyonun Kullanımı
try:
    print(ters_cevir("Python"))
    print(ters_cevir(123))  # Bu, ValueError fırlatır
except ValueError as e:
    print(e)
````

**Soru 2:** Çift Sayıları Sorgulama

Bir sayının çift olup olmadığını kontrol eden bir fonksiyon yazınız. Fonksiyonunuz, sayı çift olduğunda sayıyı geri dönsün (`return`). Eğer sayı tek ise, `ValueError` türünde bir hata fırlatsın (`raise`). Test etmek için, 20 adet hem çift hem de tek sayılar içeren bir listeyi `random` modülünü kullanarak oluşturun ve bu listede döngü kullanarak yalnızca çift sayıları `cift_sayilar.txt` isimli bir dosyaya yanyana aralarında boşluk bırakarak yazdırın.

Örneğin rastgele oluşturulan Liste: `[33, 30, 67, 88, 9, 84, 94, 64, 32, 57]` ise `cift_sayilar.txt` dosyasının içeriği aşağıdaki gibi olmalı.
```Python
Çift sayılar:
30 88 84 94 64 32
```

````python
def cift_mi(sayi):
    # Eğer sayı çift ise sayıyı geri dön
    if sayi % 2 == 0:
        return sayi
    # Eğer sayı tek ise ValueError hatası fırlat
    else:
        raise ValueError("Sayı tek.")
````

````python
cift_mi(5)
````

````python
cift_mi(6)
````

````python
import random

# 10 boyutunda rastgele tamsayılar içeren bir liste oluştur
sayilar = [random.randint(1, 100) for _ in range(10)]

print("Oluşturulan Liste:", sayilar)
````

````python
# Bu listede döngü kullanarak çift sayıları ekrana bastır
with open("cift_sayilar.txt", "w") as file:
  file.write('Çift sayılar:\n')
for sayi in sayilar:
    try:
        cift_sayi = cift_mi(sayi)
        with open("cift_sayilar.txt", "a") as file:
          file.write(f'{sayi} ')
    except ValueError as e:
        # Tek sayı olduğunda hata mesajını göz ardı et
        continue
````

**Soru 3:** Tam sayı içeren stringleri Filtreleme

Aşağıda verilen örnek liste içerisinde farklı veri tipinde değerler bulunmaktadır:

```python
liste = ["1", 2, "-5", "Python", 3.14 "2024a", "76",True, "Ahmet"]
```

Bu liste üzerinde bir döngü yardımıyla, içerisinde yalnızca pozitif tam sayı bulunan stringleri ekrana yazdırmanız gerekmektedir. Bu işlemi gerçekleştirirken, dönüşüm sırasında muhtemel hataları ele alabilmek için `try` ve `except` bloklarını kullanmayı unutmayın.

**Beklenen Ekran Çıktısı:**
```Python
1. Pozitif tamsayı: 1
Hata: 2 elemanı string değil
Hata: -5 sayısı pozitif değil
Hata: invalid literal for int() with base 10: 'Python'
Hata: 3.14 elemanı string değil
Hata: invalid literal for int() with base 10: '2024a'
2. Pozitif tamsayı: 76
Hata: True elemanı string değil
Hata: invalid literal for int() with base 10: 'Ahmet'
```

````python
liste = ["1", 2, "-5", "Python", 3.14, "2024a", "76", True, "Ahmet"]
index = 1
for item in liste:
    try:
        # Sadece string tipindeki elemanları ele al
        if isinstance(item, str):
            # String elemanı int'e çevirmeyi dene
            num = int(item)
            # Eğer int'e çevrilebilirse ama pozitif değilse özel hata yükselt
            if num <= 0:
                index += 1
                raise ValueError(f"{num} sayısı pozitif değil")
            # Eğer hiçbir hata alınmazsa ve sayı pozitifse yazdır
            print(f'{index}. Pozitif tamsayı: {item}')
        else:
            raise ValueError(f"{item} elemanı string değil")
    except ValueError as e:
        # Hata mesajını yazdır
        print(f"Hata: {e}")
````
