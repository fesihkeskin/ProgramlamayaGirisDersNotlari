**Soru 1:**

Bir mağazada, belirli bir ürünün satış fiyatı günlük olarak otomatik bir şekilde güncellenmektedir. Fiyatlandırma stratejisi, her gün ürünün fiyatını, önceki günün fiyatına göre belirlenen bir yüzde oranında artırma şeklindedir. Örneğin, eğer ürünün başlangıç fiyatı 100 TL ise ve günlük artış oranı %10 olarak belirlenmişse, ikinci gün için fiyat 110 TL, üçüncü gün 121 TL olur, bu şekilde devam eder.

Bir müşteri, ilgilendiği bu ürünü satın almayı planlamakta ve buna ne kadar bütçe ayırabileceğini önceden belirlemiştir. Sizin göreviniz, müşterinin belirlemiş olduğu bu maksimum harcama limitini aşmadan ürünü en geç kaç gün sonra satın alabileceğini hesaplayacak bir program geliştirmektir.

**Girdiler:**
- İlk günün fiyatı (int veya float).
- Günlük fiyat artış yüzdesi (int veya float).
- Müşterinin maksimum harcayabileceği miktar (int veya float).

**Çıktı:**
- Müşterinin ürünü satın alabileceği gün sayısı(int veya float).

**Beklenen Örnek Ekran Çıktısı:**
```
İlk gün fiyatını giriniz (TL): 100
Günlük artış yüzdesini giriniz (%): 10
Maksimum harcayabileceğiniz miktarı giriniz (TL):130
1. gün fiyatı: 100.00
2. gün fiyatı: 110.00
3. gün fiyatı: 121.00
Müşteri ürünü 3 gün içinde satın alabilir.
```

````python
# Kullanıcı girdilerini alın
ilk_gun_fiyati = float(input("İlk gün fiyatını giriniz (TL): "))
artis_yuzdesi = float(input("Günlük artış yüzdesini giriniz (%): "))
maksimum_harcama = float(input("Maksimum harcayabileceğiniz miktarı giriniz (TL):"))

# Gün sayısı ve mevcut fiyatı hesaplamak için değişkenleri başlatın
gun = 0
mevcut_fiyat = ilk_gun_fiyati

# Müşterinin maksimum harcama miktarını aşana kadar döngüyü sürdürün
while mevcut_fiyat <= maksimum_harcama:
    print(f"{gun+1}. gün fiyatı: {mevcut_fiyat:.2f}")
    gun += 1
    mevcut_fiyat += mevcut_fiyat * (artis_yuzdesi / 100)

# Müşterinin ürünü satın alabileceği gün sayısını yazdırın
print(f"Müşteri ürünü {gun} gün içinde satın alabilir.")
````

**Soru 2:** Bir okulun sınıf bazında başarılarını değerlendirme süreci üzerinde çalışıyoruz. Her sınıfın öğrencileri farklı final sınavı notları almıştır. Bu etkinlikte, sizden her bir sınıfın not ortalamasını hesaplayarak, okuldaki en başarılı sınıfı belirlemenizi istiyoruz. Ayrıca, bu sınıfın okuldaki sıralamasını da bulmanız gerekmektedir.

**Göreviniz:**
1. Kullanıcıdan, okulda bulunan toplam sınıf sayısını alın.
2. Kullanıcıdan, her bir sınıfın final notlarını, notlar arasında boşluk bırakarak girmelerini isteyin.
3. Girilen verileri `[[Sınıf 1 notları], [Sınıf 2 notları], ...]` şeklinde iç içe bir liste halinde düzenleyin.
4. Girilen bu verileri kullanarak her sınıfın ortalamasını hesaplayın.
5. En yüksek not ortalamasına sahip sınıfın ortalamasını ve bu sınıfın okuldaki sıralamasını ekrana yazdırın.

**Girdiler:**
- Okuldaki sınıfların final notlarını içeren iç içe bir liste. Örnek: `[[70, 80, 90], [85, 95, 100], [50, 65, 70]]`, burada her bir alt liste, bir sınıfın öğrenci notlarını temsil etmektedir.

**İpucu:**
Notları string formatında alırken, kullanıcının notları boşluklarla ayırdığından emin olun. Ardından, bu string'i `.split()` metodu ile bölerek bir liste oluşturun. Örneğin, kullanıcı "70 80 90" şeklinde bir girdi sağladığında, bu girdiyi `[70, 80, 90]` şeklinde bir listeye dönüştürebilirsiniz.

**Beklenen Örnek Ekran Çıktısı:**
```
Okulda kaç tane sınıf var? 3
1. sınıfın notlarını boşlukla ayırarak giriniz: 70 80 90
2. sınıfın notlarını boşlukla ayırarak giriniz: 85 95 100
3. sınıfın notlarını boşlukla ayırarak giriniz: 50 65 70
1. sınıfın not ortalaması: 80.00
2. sınıfın not ortalaması: 93.33
3. sınıfın not ortalaması: 61.67
En yüksek not ortalamasına sahip sınıfın ortalaması: 93.33, Okuldaki sıralaması: 2
```

````python
# Kullanıcıdan okuldaki sınıf sayısını al
sinif_sayisi = int(input("Okulda kaç tane sınıf var? "))

# Sınıfların notlarını saklamak için bir liste oluştur
siniflar_notlari = []
for i in range(sinif_sayisi):
    # Kullanıcıdan sınıfın notlarını al
    sinif_notlari = input(f"{i+1}. sınıfın notlarını boşlukla ayırarak giriniz: ").split()
    # Notları tam sayı listesine dönüştür
    sinif_notlari = [int(notu) for notu in sinif_notlari]
    # Not listesini ana listeye ekle
    siniflar_notlari.append(sinif_notlari)

# En yüksek not ortalamasını ve sıralamasını hesapla
en_yuksek_ortalama = 0
sinif_sirasi = 0

for i, sınıf in enumerate(siniflar_notlari, start=1):
    # Sınıfın not ortalamasını hesapla
    sinif_ortalama = sum(sınıf) / len(sınıf)
    # Ortalamayı yazdır
    print(f"{i}. sınıfın not ortalaması: {sinif_ortalama:.2f}")
    # En yüksek ortalama bulunduysa güncelle
    if sinif_ortalama > en_yuksek_ortalama:
        en_yuksek_ortalama = sinif_ortalama
        sinif_sirasi = i

# En yüksek ortalama ve sıralamayı yazdır
print(f"En yüksek not ortalamasına sahip sınıfın ortalaması: {en_yuksek_ortalama:.2f}, Okuldaki sıralaması: {sinif_sirasi}")
````

**Soru 3:** Bir mağazanın haftalık performans analizini yapmaktayız. Mağazada satılan farklı ürünlerin her gün ne kadar satıldığını kaydediyoruz. Bu etkinlikte, sizden hafta boyunca her bir ürün için yapılan toplam satış miktarlarını hesaplayarak, en fazla satış yapılan ürünü ve bu ürünün sırasını belirlemenizi istiyoruz.

**Göreviniz:**
1. Kullanıcıdan mağazada satılan toplam ürün sayısını alın.
2. Her bir ürün için, bir hafta boyunca her gün yapılan satış miktarlarını, satışlar arasında boşluk bırakarak girmelerini isteyin.
3. Toplam satışları ve en fazla satış yapılan ürünü belirleyin.

**Girdiler:**
- Mağazada satılan ürünlerin sayısı (tam sayı olarak).
- Her bir ürün için, bir hafta boyunca her gün yapılan satış miktarları (tam sayı, 7 gün için).

**Yönergeler:**
Günlük satış miktarlarını alırken, kullanıcının satış miktarlarını boşluklarla ayırdığından emin olun. Ardından, bu satış miktarlarını tam sayılara dönüştürerek bir liste oluşturun ve her ürün için toplam satış miktarını hesaplayın.

**Beklenen Örnek Ekran Çıktısı:**
```
Mağazada kaç farklı ürün satıldı? 2
1. ürün için günlük satış miktarlarını boşlukla ayırarak giriniz: 2 3 0 4 2 5 1
1. ürün için haftalık yapılan toplam satış miktarı: 17
2. ürün için günlük satış miktarlarını boşlukla ayırarak giriniz: 3 5 4 0 7 8 2
2. ürün için haftalık yapılan toplam satış miktarı: 29
En fazla satışı yapılan ürünün toplam satış miktarı: 29
Bu ürünün mağazadaki sırası: 2
```

````python
# Kullanıcıdan ürün sayısını al
urun_sayisi = int(input("Mağazada kaç farklı ürün satıldı? "))

# Her ürün için toplam satış miktarını saklamak için bir liste oluştur
toplam_satislar = []
for i in range(urun_sayisi ):
    # Kullanıcıdan ürünün günlük satış miktarlarını al
    gunluk_satislar = input(f"{i+1}. ürün için günlük satış miktarlarını boşlukla ayırarak giriniz: ").split()
    # Satış miktarlarını tam sayıya dönüştür ve topla
    gunluk_satislar = [int(satış) for satış in gunluk_satislar ]
    toplam_satislar.append(sum(gunluk_satislar ))
    print(f"{i+1}. ürün için haftalık yapılan toplam satış miktarı: {toplam_satislar [i]}")

# En fazla satış yapılan ürünü ve sırasını bul
en_fazla_satis = max(toplam_satislar )
en_fazla_satis_yapan_urun_sirasi = toplam_satislar.index(en_fazla_satis ) + 1

# En fazla satış yapılan ürünün toplam satış miktarını ve sırasını yazdır
print(f"En fazla satışı yapılan ürünün toplam satış miktarı: {en_fazla_satis }")
print(f"Bu ürünün mağazadaki sırası: {en_fazla_satis_yapan_urun_sirasi }")
````

**Soru 4:**

Bir sayının mükemmel olup olmadığını belirleyen `mukemmel_mi(sayi)` isimli bir fonksiyon yazın. Bir sayı, bölenlerinin toplamı kendine eşitse mükemmel sayı olarak kabul edilir. Örneğin, 6 mükemmel bir sayıdır (1 + 2 + 3 = 6). Sonra bu fonksiyonu kullanarak, 1 ile 1000 arasındaki tüm mükemmel sayıları ekrana yazdırın.

***Beklenen Ekran Çıktısı:***
```
6
28
496
```

````python
def mukemmel_mi(sayi):
  toplam=0
  for i in range(1,sayi):
    if sayi % i == 0:
      toplam += i
  return toplam == sayi
````

````python
mukemmel_mi(6)
````

````python
for sayi in range(1,1001):
  if mukemmel_mi(sayi):
    print(sayi)
````

**Soru 5:** Verilen bir sayının tüm tam bölenlerini bulan `tam_bolenleri_bul(sayi)` isimli bir fonksiyon yazınız. Ardından, bu fonksiyonu kullanarak, kullanıcıdan sürekli sayılar alın ve her bir sayının tam bölenlerini ekrana yazdırın. Kullanıcı sayı dışında bir şey girerse, program "Lütfen pozitif bir tam sayı giriniz. Program Sonlandı." mesajını yazdırarak sonlanmalıdır. Program, kullanıcı sayısal bir değer girdiği sürece çalışmaya devam etmelidir.

Programınızı yazarken, aşağıdaki adımları takip edin:

1. `tam_bolenleri_bul(sayi)` fonksiyonunu yazın. Bu fonksiyon, parametre olarak aldığı sayının tüm tam bölenlerini bir liste olarak döndürmelidir.
2. Programınızda sonsuz bir döngü oluşturun (`while True` kullanarak).
3. Döngü içinde, kullanıcıdan `input()` fonksiyonu ile bir girdi alın ve bu girdiyi bir değişkene atayın.
4. Alınan girdinin sayı olup olmadığını `.isdigit()` metodu kullanılarak kontrol edin.
5. Girdi bir sayı ise, bu sayıyı tam sayıya çevirin ve `tam_bolenleri_bul(sayi)` fonksiyonunu çağırarak sayının tam bölenlerini bulun. Bulunan tam bölenleri ekrana yazdırın.
6. Girdi bir sayı değilse, program "Lütfen pozitif bir tam sayı giriniz. Program Sonlandı." mesajını yazdırıp sonlanmalıdır.

**Beklenen Örnek Ekran Çıktısı:**
```
Sayı: 4
Tam bölenler: [1, 2, 4]
Sayı: 11
Tam bölenler: [1, 11]
Sayı: 12
Tam bölenler: [1, 2, 3, 4, 6, 12]
Sayı: a
Lütfen pozitif bir tam sayı giriniz. Program Sonlandı.
```

````python
def tam_bolenleri_bul(sayi):
    # Tam bölenlerin saklanacağı liste
    bolenler = []

    # 1'den sayıya kadar olan tüm sayılar için kontrol
    for i in range(1, sayi + 1):
        # Eğer sayı, i'ye tam bölünüyorsa, i bir bölenidir
        if sayi % i == 0:
            bolenler.append(i)

    return bolenler

# Sonsuz döngü içinde sayı girişi ve tam bölenlerin yazdırılması
while True:
    sayi_girisi = input("Sayı: ")  # Kullanıcıdan sayı girişi alınır

    if not sayi_girisi.isdigit():  # Girişin sayı olup olmadığını kontrol et
        print("Lütfen pozitif bir tam sayı giriniz. Program Sonlandı.")
        break  # Döngüden çık ve programı sonlandır
    else:
        sayi = int(sayi_girisi)  # Girişi tamsayıya çevir
        bolenler = tam_bolenleri_bul(sayi)  # Tam bölenleri bul
        print("Tam bölenler:", bolenler)  # Bölenleri yazdır
````

````python
# while'ın yanında Walrus operatörü `:=` kullanalım
while sayi_girisi := input("Sayı: "): # Sonsuz döngü içinde Kullanıcıdan sayı girişi alınır
    if not sayi_girisi.isdigit():  # Girişin sayı olup olmadığını kontrol et
        print("Lütfen pozitif bir tam sayı giriniz. Program Sonlandı.")
        break  # Döngüden çık ve programı sonlandır
    else:
        sayi = int(sayi_girisi)  # Girişi tamsayıya çevir
        bolenler = tam_bolenleri_bul(sayi)  # Tam bölenleri bul
        print("Tam bölenler:", bolenler)  # Bölenleri yazdır
````
