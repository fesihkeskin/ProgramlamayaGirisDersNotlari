# Google Colab

## 1. Google Colab Nedir?

Google Colab, tarayıcı üzerinden Python programlama dilini kullanarak interaktif şekilde kod yazmanıza, çalıştırmanıza ve paylaşmanıza olanak tanıyan ücretsiz bir bulut hizmetidir. Jupyter Notebook altyapısını kullanır ve makine öğrenimi, veri analizi ve eğitim projeleri için mükemmel bir ortamdır.

## 2. Colab'ın Avantajları

- Hiçbir kurulum gerektirmez, tarayıcınız üzerinden hemen kod yazmaya başlayabilirsiniz.
- Google Drive ile entegrasyon sağlar, bu sayede dosyalarınızı kolayca saklayabilir ve paylaşabilirsiniz.
- Ücretsiz GPU ve TPU erişimi sunar, böylece büyük veri setleri üzerinde hızlı bir şekilde çalışabilirsiniz.
- Python kitaplıkları ön yüklüdür, ek kurulumlara gerek kalmaz.

## 3. Colab'ı Kullanmaya Başlamak

### 3.1. Google Hesabınızla Giriş Yapın

Google Colab kullanabilmek için bir Google hesabınızın olması gerekmektedir. Tarayıcınızda "Google Colab" araması yaparak [Colab web sitesine](https://colab.research.google.com/) gidin ve Google hesabınızla oturum açın.

### 3.2. Yeni Bir Notebook Oluşturun

- Colab'ın ana sayfasında "New Notebook" butonuna tıklayarak yeni bir notebook oluşturabilirsiniz.
- Alternatif olarak, Google Drive'ınızda bir klasöre sağ tıklayıp "More > Google Colaboratory" seçeneğini kullanarak da yeni bir notebook oluşturabilirsiniz.

### 3.3. Colab Arayüzüne Genel Bakış

- **Menü Çubuğu:** Dosya, Düzen, Görünüm, Çalıştır, Araçlar gibi çeşitli menü seçenekleri içerir.
- **Araç Çubukları:** Notebook'u çalıştırma, kaydetme ve hücre ekleme gibi işlemleri yapmanızı sağlar.
- **Kod ve Metin Hücreleri:** Kodlarınızı yazabileceğiniz ve açıklamalar ekleyebileceğiniz bölümler.

### 3.4. İlk Kodunu Çalıştır

Aşağıdaki örnek kodu bir kod hücresine yazın ve çalıştırmak için hücrenin sol tarafındaki oynat butonuna tıklayın veya `Ctrl+Enter` tuşlarına basın:

````python
print("Merhaba Dünya!")
````

Bu komut, ekranınıza "Merhaba Colab!" mesajını yazdıracaktır.

## 4. Kaydetme ve Paylaşma

- Çalışmalarınız otomatik olarak Google Drive'ınıza kaydedilir.
- "File" menüsünden "Save a copy in Drive" seçeneği ile bir kopyasını drive'ınıza kaydedebilirsiniz.
- "Share" butonu ile notebook'unuzu diğerleriyle paylaşabilir ve izinler ayarlayabilirsiniz.

## 5. Kısayollar

**Google Colab'da kullanılan yaygın kısayol tuşları aşağıdaki gibidir:**

1. **Ctrl + Enter**: Seçili hücreyi çalıştırır.
2. **Shift + Enter**: Seçili hücreyi çalıştırır ve sonraki hücreye geçer. Eğer sonraki hücre yoksa yeni bir hücre oluşturur.
3. **Alt + Enter**: Seçili hücreyi çalıştırır ve altına yeni bir hücre ekler.
4. **Ctrl + M B**: Aktif hücrenin altına yeni bir kod hücresi ekler.
5. **Ctrl + M A**: Aktif hücrenin üstüne yeni bir kod hücresi ekler.
6. **Ctrl + M D**: Seçili hücreyi siler.
7. **Ctrl + M Z**: Son yapılan işlemi geri alır.
8. **Ctrl + M Y**: Hücre tipini kod hücresi olarak değiştirir.
9. **Ctrl + M M**: Hücre tipini metin hücresi olarak değiştirir.
10. **Ctrl + M I**: Çalışan hücreyi durdurur.
11. **Ctrl + M H**: Tüm kısayol tuşlarının listesini gösteren bir yardım penceresi açar.
12. **Ctrl + /**: Seçili satır veya satırları yorum satırı yapar ya da yorum satırını kaldırır.
13. **Ctrl + F**: Bul ve değiştir işlemleri için arama kutusunu açar.
14. **Ctrl + S**: Not defterini kaydeder.

# Python Kurulum Rehberi

## Windows İşletim Sistemi için Python Kurulumu

1. **Python'ı İndirin:**
   - Python'un resmi web sitesi olan [python.org](https://www.python.org/) üzerinden en güncel Python sürümünü indirin.
   - "Downloads" sekmesine gidin ve Windows için uygun olan sürümü seçin.

2. **Kurulumu Başlatın:**
   - İndirdiğiniz `.exe` dosyasını çalıştırın.
   - "Install Python 3.x" penceresinde, en altta bulunan "Add Python 3.x to PATH" seçeneğini işaretleyin. Bu adım, Python'u ve pip'i sistem yoluna ekler.
   - "Install Now" butonuna tıklayarak kurulumu başlatın.

3. **Kurulumu Kontrol Et:**
   - Kurulum bittikten sonra, komut istemcisini (Command Prompt) açın ve `python --version` komutu ile Python sürümünü kontrol edin. Kurulan sürüm numarasını görmelisiniz.

## macOS İşletim Sistemi için Python Kurulumu

1. **Homebrew Kullanarak Python Kurulumu:**
   - Terminal'i açın.
   - Eğer Homebrew yüklü değilse, aşağıdaki komutu kullanarak yükleyin:
     ```
     /bin/bash -c "$(curl -fsSL https://raw.githubusercontent.com/Homebrew/install/HEAD/install.sh)"
     ```
   - Homebrew yüklüyse, aşağıdaki komut ile Python'u kurun:
     ```
     brew install python
     ```

2. **Kurulumu Kontrol Et:**
   - Kurulum tamamlandıktan sonra, `python3 --version` komutu ile Python sürümünü kontrol edin.

## Linux İşletim Sistemi için Python Kurulumu

1. **Python Kurulumu:**
   - Çoğu Linux dağıtımı Python ile birlikte gelir. Ancak güncel bir sürüm yüklemek istiyorsanız, terminal üzerinden aşağıdaki komutları kullanabilirsiniz.
   - Debian tabanlı sistemler (Ubuntu gibi) için:
     ```
     sudo apt-get update
     sudo apt-get install python3
     ```
   - Red Hat tabanlı sistemler için:
     ```
     sudo yum install python3
     ```

2. **Kurulumu Kontrol Et:**
  
  2.1. **Command Prompt (CMD) veya PowerShell:**
    - CMD veya PowerShell'i açın.
    - Komut satırına `python --version` yazıp Enter'a basarak Python sürümünü kontrol edin.

  2.2. **Python IDLE:**
    - Python IDLE'ı açın.
    - IDLE'ın üst kısmında gösterilen veya interaktif kabukta `import sys; print(sys.version)` komutunu çalıştırarak Python sürümünü kontrol edin.

# VS Code yükleme

#### 1. Giriş
Visual Studio Code (VS Code), Microsoft tarafından geliştirilen ücretsiz, açık kaynaklı bir kod editörüdür. JavaScript, TypeScript ve Node.js ile birlikte pek çok dil için destek sunar ve geniş bir eklenti yelpazesiyle özelleştirilebilir.

#### 2. Kurulum Öncesi Gereklilikler
- İnternet bağlantısı
- İşletim sisteminin güncel olması önerilir (Windows, macOS, Linux)

#### 3. Kurulum Adımları

##### Windows:
1. Microsoft'un resmi web sitesine gidin ([Visual Studio Code indirme sayfası](https://code.visualstudio.com/Download)).
2. Windows için uygun olan 'User Installer' sürümünü indirin.
3. İndirilen kurulum dosyasını çalıştırın.
4. Kurulum sihirbazında, lisans anlaşmasını kabul edin ve kurulum için talimatları izleyin.
5. Kurulum tamamlandıktan sonra VS Code'u başlatın.

##### macOS:
1. Microsoft'un resmi web sitesine gidin ([Visual Studio Code indirme sayfası](https://code.visualstudio.com/Download)).
2. macOS için uygun olan sürümü indirin.
3. İndirilen '.zip' dosyasını açın ve 'Visual Studio Code' uygulamasını 'Applications' klasörüne sürükleyin.
4. Uygulamayı 'Applications' klasöründen açın.

##### Linux:
1. Microsoft'un resmi web sitesine gidin ([Visual Studio Code indirme sayfası](https://code.visualstudio.com/Download)).
2. Linux dağıtımınıza uygun olan sürümü indirin (.deb veya .rpm).
3. Terminal üzerinden, indirilen paket dosyası için uygun kurulum komutunu kullanın. Örneğin, Debian/Ubuntu için:
   ```
   sudo dpkg -i <dosya_adı.deb>
   sudo apt-get install -f
   ```
4. Kurulum tamamlandıktan sonra, uygulamayı uygulama menüsünden başlatın.

#### 4. İlk Ayarlar ve Eklentilerin Yüklenmesi
Kurulumdan sonra, kişiselleştirmek ve programlama diliniz için gerekli araçları kurmak amacıyla Extensions (Eklentiler) bölümünden istediğiniz eklentileri aratabilir ve yükleyebilirsiniz. Örneğin, Python geliştiricileri için 'Python', JavaScript geliştiricileri için 'ESLint' eklentileri önerilir.

- Visual Studio Code, genişletilebilir yapısı ve desteklediği dillerin çeşitliliği ile modern programlama ihtiyaçlarınız için mükemmel bir araçtır. Kurulumu basit ve kullanımı kolaydır. Bu ders notları ile VS Code kurulumunu başarıyla gerçekleştirebilir ve kodlama çalışmalarınıza hemen başlayabilirsiniz.

# **VS Code Kurulum ve Hazırlama**

Geliştirme ortamınızı kurma ve ayarlama sürecini aşama aşama yeniden ele alalım, eklemelerle birlikte:

1. **VS Code'u Kurun**: [Visual Studio Code'un resmi web sitesinden](https://code.visualstudio.com/) işletim sisteminize uygun yükleme dosyasını indirin ve kurulumu gerçekleştirin. VSCode'u açtıktan sonra soldaki panelden Eklentiler `Extensions` tıklayıp Python'u yükleyin.

2. Bilgisayarınızda birden fazla Python sürümü var ise,
  - **Command Palette'i Açın**: `Ctrl + Shift + P` (Windows/Linux) veya `Cmd + Shift + P` (Mac) tuşlarına basarak açabilirsiniz.
  - **Python Interpreter'ı Seçin**: Command Palette'e `Python: Select Interpreter` yazın ve çıkan seçeneklerden bu komutu seçin.
  - **Sürümü Seçin**: Listeden uygun Python sürümünü seçin. Eğer kullanmak istediğiniz sürüm listede yoksa, bu Python sürümünün kurulu olduğundan emin olun ve kurulum yolunu manuel olarak ekleyin.

3. **Proje Klasörü Oluşturun ve VS Code’da Açın**: Projeniz için bir klasör oluşturun ve VS Code'da `File` menüsünden `Open Folder` seçeneği ile bu klasörü açın. Ya da, windows'ta sağ tık `Open with Code` tıklayarak klasörü açın. Ya da windows'taki bir klasörü VS Code'un solundaki Explorer'in içine sürükle bırak yapın.

4. **Workspace Oluşturun**: Açılan proje klasörünü `File` > `Save Workspace As...` ile bir çalışma alanı olarak kaydedin.

5. **Virtual Enviroment (Sanal Ortam) Kurun**: VS Code'da, terminali açmak için `View` menüsünden `Terminal` seçeneğini veya `Ctrl + "` kısayolunu kullanın. Sanal ortam kurmadan önce terminale `python --version` yazarak versiyon kontrolü yapın. Değişiklik yapacaksanız 2. maddeyi tekrarlayın. Daha sonra terminale `python -m venv <sanal_ortam_adı>` yazarak sanal ortamı oluşturun. `<klasor_adı> (WORKSPACE)` altında sanal ortamınızın adı ile aynı `<sanal_ortam_adı>` klasörü oluşacaktır. Bu klasörün içinde varsayılan bazı klasörler ve `.cfg` uzantılı sanal ortamınızın konfigürasyon dosyası bulunmaktadır. Oluşturulan sanal dosyayı aktifleştirmek için terminale `.\<sanal_ortam_adı>\Scripts\activate` yazın. Eğer `powershell`'de hata alırsanız `cmd`'de de tekrar deneyin. **Not:** `.py` uzantılı bir dosyanın içindeyken sağ alt durum çubuğunda `Python 3.12.1 ('<sanal_ortam_adı>':venv)` göreceksiniz. Buradan hangi Python sürümü ve sanal ortamda olduğunuzu kontrol edebilirsiniz. Sanal ortam oluşturulduktan ve aktive edildikten sonra kullanacağınız kütüphane ve modülleri pip aracılığıyla terminal üzerinden yükleyiniz.

4. **Dosya ve Klasör Yapısını Kurun**: Explorer panelinden proje dosya ve klasör yapısını oluşturun.

5. **Git Reposu Başlatın** (isteğe bağlı): Terminalden `git init` komutu ile Git reposunu başlatın ve Source Control özelliklerini kullanın.

6. **Eklentileri Yükleyin**:
   - **Python Eklentileri**: `Extensions` panelinden Python için IntelliCode, Django gibi eklentileri, ve Pylance'ı yükleyin ki bu da Python ile kod yazarken size güçlü bir IntelliSense deneyimi sunar.

   - **Jupyter**: Eğer Jupyter Notebook'larla çalışmak istiyorsanız, Colab'dan indirdiğiniz dosyaları da çalıştırabilmenizi sağlayacak Jupyter eklentisini yükleyin. Normal `.py` uzantılı bir Python dosyasında kodları mouse ile seçip `Shift+Enter` tuş kombinasyonu ile sadece seçili kodları Jupyter interactive ile çalıştırabilirsiniz. Bunun için sol alt köşedeki ayarlar simgesine sonra `Settings` (ya da `Ctrl+,`) tıklayıp gelen arama kutusuna 'jupyter' yazarak `Jupyter: Enable Kernel Completions` seçeneğini etkinleştirin.

   - **Temalar ve İkonlar**: (Tavsiyem) Material Icon Theme'i kurun ve dosyalar arasında daha rahat gezinmek için `Material Icon Theme`'i tercih edin. Bunun için sol alt köşedeki ayarlar simgesine tıklayıp, `Themes` -> `File Icon Theme`-> `Material Icon Theme` seçeneklerinden yapabilirsiniz.

   - **gitlens**


7. **.ipynb ve .py Uzantılarını Kullanın**: `.ipynb` uzantılı Jupyter notebook dosyalarında hücre bazlı kod çalıştırabilirsiniz. `.py` dosyaları ise genellikle top-down (baştan-sona) çalışır, fakat sol alttan ayarlar tıklayıp `Jupyter: Send Selection to Interactive Window` aratıp özelliği aktif yaparsanız, `.py` uzantılı dosyalarda da belirli kod bloklarını (imlecin bulunduğu satırı ya da seçili kodları) shift+enter ile çalıştırabilirsiniz.

8. **Ortam Ayarlarını Yapın**: Projeye özgü çevre ayarlarını, örneğin Python yorumlayıcı seçimi veya sanal ortamlar (virtual environments), ayarlamak için sol alt köşeden ayarlar simgesine tıklayın ve ilgili ayarları arayın veya direkt `settings.json` dosyasını düzenleyin.

Bu aşamalardan sonra kod yazmaya başlayabilir ve proje geliştirmeye devam edebilirsiniz. Her bir özellik ve ayar, geliştirme sürecinizi kişiselleştirmenizi ve optimize etmenizi sağlar.
