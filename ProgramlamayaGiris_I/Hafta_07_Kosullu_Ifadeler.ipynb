Python, matematiğin olağan mantıksal koşullarını destekler:

Eşittir: a == b

Eşit Değil: a != b

Şundan daha az: a < b

Şundan küçük veya eşit: a <= b

Şundan büyük: a > b

Büyük veya eşit: a >= b

**elif** anahtar sözcüğü Python'un "eğer önceki koşullar doğru değilse, o zaman bu koşulu dene" deme şeklidir.

```python
a = 100
b = 15
if b > a:
  print("b, a'dan büyüktür.")
elif a == b:
  print("a, b'ye eşittir.")
```

**else** anahtar sözcüğü, önceki koşullar tarafından yakalanmayan her şeyi yakalar.

```python
a = 15
b = 15
if b > a:
  print("b, a'dan büyüktür.")
elif a == b:
  print("a, b'ye eşittir.")
else:
  print("a, b'den büyüktür.")
```

**elif** olmadan **else** de kullanabilirsiniz:

```python
a = 100
b = 15
if b > a:
  print("b, a'dan büyüktür.")
else:
  print("b, a'dan büyük değildir.")
```

**if**'in kısa yazım şekli: Yürütülecek yalnızca bir deyiminiz varsa, bunu **if** deyimiyle aynı satıra koyabilirsiniz.

```python
a = 15
b = 100
if b > a:
  print("b, a'dan büyüktür.")
```

```python
a = 15
b = 150
if b > a: print("b, a'dan büyüktür.")
```

Yürütülecek yalnızca bir deyiminiz varsa, bir **if** ve bir **else** için, hepsini aynı satıra koyabilirsiniz.

```python
a = 100
b = 15
if b > a:
  print("b, a'dan büyüktür.")
else:
  print("b, a'dan büyük değildir.")
```

```python
a = 10
b = 15
print("b, a'dan büyüktür.") if b > a else print("b, a'dan büyük değildir.")
```

Tek satırda birden fazla **else** ifadesi de kullanabilirsiniz

```python
a = 100
b = 100
print("a, b'den büyüktür.") if a > b else print("a ve b eşittir") if a==b else print("b, a'dan büyüktür.")
```

```python
if a > b:
  print("a, b'den büyüktür.")
else:
  if a==b:
    print("a ve b eşittir")
  else:
      print("b, a'dan büyüktür.")
```

**and** anahtar sözcüğü mantıksal bir işleçtir ve koşullu ifadelerin aynı anda gerçekleşmesi durumudur.

```python
a=100
b=15
c=200

if a>b and c>a:
  print("koşulların ikisi de doğru")
```

**or** anahtar sözcüğü mantıksal bir işleçtir ve koşullu ifadelerden birinin gerçekleşmesi yeterlidir.

```python
a=100
b=150
c=90

if a>b or c>a:
  print("koşullardan en az biri doğru")
```

**not** anahtar sözcüğü mantıksal bir işleçtir ve koşullu ifadenin sonucunu tersine çevirmek için kullanılır:

```python
a=100
b=15

if not a<b:
  print("b, a'dan büyük DEĞİL'dir")
```

```python
a=100
b=15

if a>b:
  print("b, a'dan büyük DEĞİL'dir")
```

# İç içe if koşulu ( Nested if)

İf deyimlerinin içinde if deyimleri olabilir, buna iç içe if (Nested if) deyimleri denir.


```python
x=15

if x > 10:
  print("Girilen x değeri 10'dan büyüktür")
  if x>20:
    print("Girilen x değeri 20'den de büyüktür")
  else:
    print("Fakat 20'den büyük değildir")
else:
  print("Girilen x değeri 10'dan küçüktür")
```

**if** deyimleri boş olamaz, ancak herhangi bir nedenle içeriği olmayan bir if deyiminiz varsa, hata almamak için **pass** deyimini ekleyin.

```python
a=15
b=150

if a<b:
  pass
```

2 ve/veya 3'e Bölünebilirlik Kontrolü:

```python
sayi = int(input("Bir sayı giriniz:"))

if sayi % 2 == 0 and sayi % 3 == 0:
  print("Sayı 2 ve 3'e tam bölünür")
elif sayi % 2 == 0:
  print("Sayı 2'ye tam bölünür")
elif sayi % 3 == 0:
  print("Sayı 3'e tam bölünür")
else:
  print("Sayı 2 ve 3'e de tam bölünmez")
```
