Alıştırma 4.1:Kullanıcıdan bir odanın genişliğini ve uzunluğunu girmesini isteyen bir program yazınız. Kullanıcı değerleri girdikten sonra, programınız odanın alanını hesaplamalı ve ekrana yazdırmalıdır. Uzunluk ve genişlik değişkenleri float veri türü olarak girilecektir.

```python
# Kullanıcıdan oda uzunluğu ve genişliği girmesini istiyoruz
uzunluk = float(input("Lütfen odanın uzunluğunu giriniz (metre cinsinden): "))
genislik = float(input("Lütfen odanın genişliğini giriniz (metre cinsinden): "))

# Alanı hesaplıyoruz
alan = uzunluk * genislik

# Sonucu ekrana yazdırıyoruz
print(f"Odanın alanı: {alan} metrekare")
```

**Alıştırma 4.2:**
Kullanıcıdan a ve b olmak üzere iki **tamsayı** girmesini isteyen bir program yazınız. Programınız girilen iki sayının toplamını, çıkarmasını, çarpmasını ve bölmesini hesaplamalı ve ekrana yazdırmalıdır.

```python
# Kullanıcıdan a ve b tamsayılarını girmesini istiyoruz
a = int(input("Lütfen 'a' tamsayısını giriniz: "))
b = int(input("Lütfen 'b' tamsayısını giriniz: "))

# Aritmetik işlemleri gerçekleştiriyoruz
toplam = a + b
cikarma = a - b
carpma = a * b
bölme = a / b

# Sonuçları ekrana yazdırıyoruz
print(f"{a} + {b} = {toplam}")
print(f"{a} - {b} = {cikarma}")
print(f"{a} x {b} = {carpma}")
print(f"{a} / {b} = {bölme}")
```

**Alıştırma 4.3:**
Kullanıcıdan tam sayı olan bir yarıçap (r) değeri girmesini isteyen ve r yarıçaplı bir dairenin alanını ile r yarıçaplı bir kürenin hacmini hesaplayarak ekrana yazdıran programı yazınız. Pi sayısını 3.14 alabilirsiniz.

```python
# Kullanıcıdan tam sayı olan yarıçap değerini girmesini istiyoruz
r = int(input("Lütfen dairenin ve kürenin yarıçapını (r) tam sayı olarak giriniz: "))

# Sabit olarak pi değerini tanımlıyoruz
pi = 3.14

# Dairenin alanını ve kürenin hacmini hesaplıyoruz
dairenin_alani = pi * r * r
kurenin_hacmi = (4/3) * pi * r * r * r

# Sonuçları ekrana yazdırıyoruz
print(f"r = {r} yarıçaplı dairenin alanı: {dairenin_alani} metrekare")
print(f"r = {r} yarıçaplı kürenin hacmi: {kurenin_hacmi} metreküp")
```

**Alıştırma 4.4:**

**Alıştırma 4.5:**
a ve b kenar uzunlukları kullanıcı tarafından girilen bir dik üçgenin hipotenüs uzunluğunu hesaplayan ve ekrana yazdıran programı yazınız.

```python
# Kullanıcıdan a ve b kenar uzunluklarını alıyoruz
a = float(input("Lütfen dik üçgenin 'a' kenar uzunluğunu giriniz: "))
b = float(input("Lütfen dik üçgenin 'b' kenar uzunluğunu giriniz: "))

# Hipotenüs uzunluğunu hesaplıyoruz (Pisagor Teoremi kullanılarak)
hipotenüs = (a**2 + b**2)**0.5

# Sonucu ekrana yazdırıyoruz
print(f"a = {a} ve b = {b} kenar uzunluklarına sahip dik üçgenin hipotenüs uzunluğu: {hipotenüs}")
```
