## While Döngüleri (While Loops)

```
while (koşul doğru):
    yapılacak işlemler
```



while döngüsü ile bir koşul doğru (True) olduğu sürece bir dizi deyimi çalıştırabiliriz.

````python
# sayac değeri'ni 5'den küçük olduğu sürece ekrana yazdır:
sayac = 0
while sayac < 5:
    print("Sayac değeri:", sayac)
    sayac += 1

# Not: i'yi artırmayı unutmayın, yoksa döngü sonsuza kadar devam eder.
# Sonsuz döngüye girmemek için koşulun bir noktada False olması gerekir.
# while döngüsünde, ilgili değişkenlerin hazır olması gerekir,
# bu örnekte, sayac = 0 olarak belirlediğimiz bir indeksleme değişkenini tanımladık.
````

*Yukarıdaki kod parçacığının adımları aşağıdaki gibidir:*

**Sayacın Başlatılması:**

İlk olarak, `sayac` adında bir değişken tanımlıyoruz ve ona `0` değerini atıyoruz. Bu değişken, döngümüzün kaç kez çalıştığını takip edecektir. Döngü her tekrarlandığında bu değişkenin değerini artırarak, döngünün bir noktada sonlanmasını sağlayacağız.

**While Döngüsünün Başlaması:**

while anahtar kelimesiyle bir döngü başlatıyoruz. Döngünün devam etme koşulu `sayac < 5 `şeklinde belirtilmiştir. Bu ifade, `sayac` değişkeninin değeri `5`'ten küçük olduğu sürece döngünün devam edeceği anlamına gelir.

**İşlem Bloğu:**

Döngünün içinde, `print` fonksiyonu ile her döngü iterasyonunda sayac değişkeninin mevcut değerini ekrana yazdırıyoruz.

**Değeri Güncelleme:**

`sayac += 1` ifadesi ile sayac değişkeninin değerini her döngüde bir artırıyoruz. Bu ifade, `sayac = sayac + 1` ifadesinin kısa yazılışıdır ve `sayac`'ın mevcut değerine `1` ekler. Bu adım, döngünün bir noktada sona ermesini garanti eder; aksi takdirde döngü sonsuza kadar devam eder (sonsuz döngü).

**Döngü Sonlandırma:**

Her iterasyonda sayac arttıkça, while `sayac < 5` koşulu bir noktada `False` (yanlış) olacaktır. `sayac` değeri `5`'e ulaştığında, koşul yanlış olur ve döngü sona erer. Döngü sona erdiğinde programın kontrolü, while döngüsünden sonraki ilk satıra geçer.

Bu döngü, `sayac` değeri `0`'dan başlayarak `4`'e kadar olan (5 dahil olmadan) sayıları ekrana yazdırır. Her bir sayıyı yazdırdıktan sonra, `sayac` bir artırılır ve `while` döngüsünün koşulu tekrar kontrol edilir. Bu süreç, koşul `False` olana kadar devam eder.

# break deyimi (break statement)

`break` deyimi, içinde bulunduğu döngünün (`while`, `for`) anında sonlandırılmasını sağlayan bir kontrol deyimidir. Yani, döngü belirtilen koşul ne olursa olsun `break` ifadesine ulaştığında döngüyü terk eder ve döngünün hemen sonrasındaki kodların çalışmasına devam edilir.


*   `break`, genellikle bir `if` (koşul) bloğu içerisinde kullanılır ve belirli bir koşulun gerçekleşmesi durumunda döngüden çıkmak için kullanılır.
*   `break` genellikle, döngünün normal bitiş koşulu dışında bir koşul gerçekleştiğinde döngüden çıkmak istediğimizde kullanılır. Örneğin, bir hata durumunda veya istenilen bir sonuca ulaşıldığında döngüyü bitirmek için `break` kullanabiliriz.






````python
i = 1
while i < 6:
  print(i)
  if i == 3:
    break
  i += 1
````

Bu kod, i değişkenini kullanarak 1'den başlayıp 6'dan küçük olana kadar (yani 5'e kadar) sayıları ekrana yazdırmak için bir while döngüsü kullanır. Ancak i değişkeninin değeri 3 olduğunda break ifadesiyle döngüyü anında sonlandırır.

# continue deyimi (continue statement)

`continue` deyimi ile mevcut iterasyonu (yinelemeyi) durdurur ve `while`'ın bir sonraki iterasyonuna geçer.

````python
i = 0
while i < 6:
  i += 1
  if i == 3:
    continue # while ile devam et bir alt satıra inme
  print(i)
# Sonuçlarda 3 rakamının eksik olduğuna dikkat edin
````

Bu kod, 1'den 6'ya kadar olan sayıları yazdırır, ancak `i` değeri 3 olduğunda sayıyı yazdırmadan döngünün başına döner.

Kodun çıktısı, `i` değeri 3 olduğunda bu sayıyı atladığı için 1, 2, 4, 5, 6 olacaktır. `i` değeri 3'e eşit olduğunda `continue` ifadesi nedeniyle `print(i)` satırı atlanır ve 3 yazdırılmaz. Döngü `i` değeri 6 olduğunda sona erer çünkü `while` döngüsünün koşulu (`i < 6`) artık `False` olur.

# else deyimi (else statement)

`else` deyimiyle, `while`'daki koşul artık doğru olmadığında `else`'in içine yazılan kod bloğunu bir kez çalıştırabiliriz.

````python
# Koşul yanlış olduğunda bir mesaj yazdırın:
i = 1
while i < 6:
  print(i)
  i += 1
else:
    print("else içindeki bu kod while'daki koşul olan \n i artık 6'dan küçük değil durumunda bir kez çalıştı")
    print("Burada çalışır")
````

## Python'da Kullanıcı Girdisi `input` Alma ve `while` ile Doğrulama

### Kullanıcıdan Veri Alma

Python'da kullanıcıdan veri almak için `input()` fonksiyonu kullanılır. Bu fonksiyon, programın kullanıcıdan bir girdi istemesini sağlar ve kullanıcının giriş yaptığı metni bir string (metin) olarak geri döndürür.

Örneğin, bir kullanıcıdan ismini isteyebiliriz:

````python
user_name = input("Lütfen adınızı girin: ")
print(f"Merhaba, {user_name}!")
````

### Girdiyi Doğrulama

Çoğu durumda, kullanıcının girdiği verinin belirli bir formatta olmasını veya belirli koşulları karşılamasını isteriz. Örneğin, bir sayının pozitif olmasını veya kullanıcı adının belirli bir uzunlukta olmasını isteyebiliriz.

Bu tür durumlar için, girdiyi doğrulamak üzere bir döngü (genellikle `while` döngüsü) kullanabiliriz. Kullanıcı doğru formatta bir girdi sağlayana kadar döngü devam eder.

**Örnek:** Pozitif Bir Sayı Alma

Kullanıcıdan pozitif bir sayı girmesini isteyelim. Kullanıcı negatif bir sayı veya bir sayı olmayan bir değer girerse, yeni bir giriş yapması için tekrar istekte bulunalım:

````python
while True:
    user_input = input("Lütfen pozitif bir sayı girin: ")
    if user_input.isdigit() and int(user_input) > 0:
        print("Teşekkürler, doğru bir değer girdiniz!")
        break  # Doğru girdi alındı, döngüden çık
    else:
        print("Hata: Lütfen pozitif bir sayı girin.")
````

Bu örnekte, `isdigit()` metodu string'in bir sayı olup olmadığını kontrol eder. Eğer bu koşul ve sayının pozitif olması koşulu sağlanırsa, döngü `break` ifadesi ile sonlandırılır.

### Alıştırmalar

2. Kullanıcıdan bir e-posta adresi girmesini isteyin ve girilen değerin "@" işaretini içerip içermediğini kontrol edin.
3. Kullanıcıdan bir parola ve parolanın tekrarını isteyin. İki parolanın da eşleştiğini doğrulayın.
