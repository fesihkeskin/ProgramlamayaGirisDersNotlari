Bir dizede belirli bir ifadenin veya karakterin bulunup bulunmadığını kontrol etmek için **in** anahtar sözcüğünü kullanabiliriz.

```python
txt = "Python ve Django iyi bir ikilidir"
print("iyi" in txt)
```

Bir dizede belirli bir ifadenin veya karakterin bulunup bulunmadığını kontrol etmek için not in anahtar sözcüğünü kullanabiliriz.

```python
txt = "Python ve Django iyi bir ikilidir"
print("iyi" not in txt)
```

String      ->  P y t h o n

indeks      ->  0 1 2 3 4 5

ters indeks -> -6-5-4-3-2-1

```python
a = "Python"
print('a =',a)
print('len(a):',len(a))
print('a[0]:',a[0])
print('a[1]:',a[1])
print('a[-1]:',a[-1])
print('a[-2]:',a[-2])
print('a[2:5]:',a[2:5])
print('a[:3]:',a[:3])
print('a[2:]:',a[2:])
print('a[0:6:2]:',a[0:6:2])
```

```python
# Kullanıcıdan bir cümle alıyoruz
cumle = input("Lütfen bir cümle giriniz: ")

# Cümlenin karakter sayısını buluyoruz
karakter_sayisi = len(cumle)

# Cümlenin ilk karakterini buluyoruz
ilk_karakter = cumle[0]

# Cümlenin son karakterini buluyoruz
son_karakter = cumle[-1]

# Cümlenin kelime sayısını buluyoruz
kelime_sayisi = len(cumle.split())

# Cümlenin 'a' harfi sayısını buluyoruz
a_sayisi = cumle.count('a')

# Sonuçları ekrana yazdırıyoruz
print(f"Girilen cümle {karakter_sayisi} karakterden oluşmaktadır.")
print(f"Girilen cümlenin ilk karakteri: {ilk_karakter}")
print(f"Girilen cümlenin son karakteri: {son_karakter}")
print(f"Girilen cümle {kelime_sayisi} kelimeden oluşmaktadır.")
print(f"Girilen cümlede {a_sayisi} tane 'a' karakteri bulunmaktadır.")
```
