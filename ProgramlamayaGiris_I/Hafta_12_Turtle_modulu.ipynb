### Turtle Modülü ile Yeni Çizimlere Başlama

Turtle modülü ile çizim yaparken, kaplumbağanın en son kaldığı konum ve yönü, bir sonraki çizim işlemi için başlangıç noktası olacaktır. Bu durumu yönetmek için aşağıdaki yöntemleri kullanabilirsiniz.

#### Yeni Bir Turtle Başlatma

Yeni bir çizim yaparken, eğer `initializeTurtle()` ve `Turtle()` komutlarını kullanmazsak, kaplumbağa en son kaldığı yerden devam eder. Bu, önceki çizimlerin konum ve yön bilgilerini taşıyabileceği anlamına gelir. Yeni bir kaplumbağa başlatmak için şu kodu kullanabilirsiniz:

````
# ColabTurtlePlus'ı turtle ismi ile içe aktaralım
import ColabTurtlePlus.Turtle as turtle

# ColabTurtlePlus'ı başlat
turtle.initializeTurtle()  # Eğer initialize (başlat) yapmazsanız en son kaldığı yerden devam eder

# Kaplumbağa objesi oluştur ve 'kalem' değişkenine ata
kalem = turtle.Turtle()
````

Bu kod, Colab ortamında turtle modülünü başlatır ve yeni bir kaplumbağa oluşturur. Ancak, her yeni çizim için bu kodu yazmak yerine, daha basit bir yöntem de kullanabilirsiniz.

### Mevcut Kaplumbağayı Sıfırlama Varsayılan Ayarlara Döndürmek

**Not:** Colab gibi Jupyter Notebook ortamlarında, turtle çizimleri bellekte kalabilir. Bu nedenle, temiz bir çizim sayfası elde edemiyorsanız, çalışma zamanı sekmesinden oturumu yeniden başlatmayı deneyebilirsiniz.

#### 1. `reset()` Fonksiyonu

`reset()` fonksiyonu, kaplumbağanın konumunu, yönünü, rengini ve tüm diğer özelliklerini varsayılan ayarlarına döndürür. Ayrıca, çizim alanını temizler.

````
# Kaplumbağayı oluştur
kalem = turtle.Turtle()

# Kaplumbağayı varsayılan ayarlarına döndür
kalem.reset()

# Bazı özellikleri ayarlayalım
kalem.pencolor("red")
kalem.shape("turtle")

kalem.forward(100)
````

````
# Kaplumbağayı varsayılan ayarlarına döndür
kalem.reset()
````

Bu fonksiyon, kaplumbağayı (ve çizim yüzeyini) sıfırlar ve kaplumbağa başlangıç konumuna geri döner. Bu sayede, yeni bir çizime başlamak için tekrar `initializeTurtle()` ve `Turtle()` komutlarını yazmanız gerekmez.

````
# Yeni bir çizim yapalım
kalem.forward(50)
````

#### 2. `clearscreen()` fonksiyonu ile sıfırlama
`turtle.clearscreen()` fonksiyonu turtle grafik ekranını temizlemek ve tüm turtle nesnelerini kaldırmak için kullanılır. Hem Jupyter Notebook hem de standart Python betikleri içinde bu fonksiyonu kullanabilirsiniz.

````
# Ekranı temizle
turtle.clearscreen()

# turtle.clearscreen() fonksiyonu kalem objesini de sildi
# Yeni bir kaplumbağa oluştur ve yeni çizimler yapalım
kalem = turtle.Turtle()
kalem.forward(50)
````

#### 3. `clear()` Fonksiyonu

`clear()` fonksiyonu sadece çizim alanını temizler, kaplumbağanın konumunu, yönünü ve diğer özelliklerini değiştirmez.

````
# Kaplumbağayı oluştur
kalem = turtle.Turtle()

# Bazı özellikleri ayarlayalım
kalem.pencolor("blue")
kalem.shape("circle")
kalem.forward(100)
````

````
# Çizim alanını temizle
kalem.clear()

# Kaplumbağa mevcut konum ve özellikleri ile çizmeye devam eder
kalem.forward(100)
````

**NOT:**

- **`clear()`**: Mevcut kaplumbağanın çizimlerini temizler, kaplumbağanın konumunu ve yönünü değiştirmez.
- **`clearscreen()`**: Tüm ekranı temizler, tüm kaplumbağaları kaldırır ve ekranı varsayılan duruma döndürür.

### Turtle Modülünde Pozitif ve Negatif Pixel ve Açı Kullanımı

Python'un `turtle` modülünde, kaplumbağayı hareket ettirirken pozitif ve negatif değerler kullanarak hareket yönünü ve açılarını kontrol edebiliriz.

#### Pozitif Pixel ve Açı Kullanımı

Pozitif değerler kullanarak kaplumbağayı ileri hareket ettirebilir ve saat yönünde döndürebiliriz.

````
# Kaplumbağayı sıfırla
kalem.reset()

# Kaplumbağayı ileri hareket ettir
kalem.forward(100)  # 100 piksel ileri gider
# Kaplumbağayı sağa döndür
kalem.right(90)  # Saat yönünde (sağa) 90 derece döner
# Kaplumbağayı tekrar ileri hareket ettir
kalem.forward(50)  # 50 piksel ileri gider
````

#### Negatif Pixel ve Açı Kullanımı

Negatif değerler kullanarak aynı hareket ve dönüşleri gerçekleştirebiliriz. Negatif değerler, pozitif değerlerin ters yönünde çalışır.

````
# Kaplumbağayı sıfırla
kalem.reset()

# Kaplumbağayı negatif değerlerle hareket ettir
kalem.backward(-100)  # -100 piksel geri gider, aslında 100 piksel ileri gider
kalem.left(-90)  # -90 derece sola döner, aslında sağa döner
kalem.backward(-50)  # -50 piksel geri gider, aslında 50 piksel ileri gider
````

Bu kod, kaplumbağanın hem pozitif hem de negatif değerler kullanarak aynı hareketleri gerçekleştirebileceğini gösterir. Negatif değerler, pozitif değerlerin ters yönünde çalışır, bu da hareketleri ve dönüşleri daha esnek bir şekilde kontrol etmenizi sağlar.

````
# Bir kare çizelim
kalem.reset()

kalem.forward(100)
kalem.right(90)

kalem.forward(100)
kalem.right(90)

kalem.forward(100)
kalem.right(90)

kalem.forward(100)
kalem.right(90)
````

````
# Bir kare çizelim
kalem.reset()

for _ in range(4):
    kalem.forward(100)
    kalem.right(90)
````

````
# Bir yuvarlak çizmek için circle() fonksiyonunu kullanıyoruz
kalem.reset()

kalem.circle(100)  # 100 piksel yarıçapında bir daire çizer
````

````
# okun son kaldığı yerden devam edip merkeze götürelim
kalem.left(90)  # 90 derece sola döner
kalem.forward(100)  # 50 ileri
````

### Çokgenlerin İç ve Dış Açıları

Bir çokgenin iç ve dış açılarını hesaplamak için bazı temel formüller kullanılır.

#### İç Açı Formülü

Bir $ n $ kenarlı düzgün çokgenin her bir iç açısını hesaplamak için kullanılan formül:

$$ \text{İç Açı} = \frac{(n-2) \times 180^\circ}{n} $$

Burada $ n $, çokgenin kenar sayısını temsil eder.


ya da

$$ \text{İç Açı} = 180^\circ -\text{Dış Açı} $$

#### Dış Açı Formülü

Bir $ n $ kenarlı düzgün çokgenin her bir dış açısını hesaplamak için kullanılan formül:

$$ \text{Dış Açı} = \frac{360^\circ}{n} $$

Burada $ n $, çokgenin kenar sayısını temsil eder.

ya da

$$ \text{Dış Açı} = 180^\circ -\text{İç Açı} $$

### Örnekler

#### Beşgen (n = 5)

- İç Açı:
  $$ \text{İç Açı} = \frac{(5-2) \times 180^\circ}{5} = \frac{3 \times 180^\circ}{5} = 108^\circ $$

- Dış Açı:
  $$ \text{Dış Açı} = \frac{360^\circ}{5} = 72^\circ $$


Bu formülleri kullanarak, herhangi bir düzgün çokgenin iç ve dış açılarını kolayca hesaplayabilirsiniz.

**Örnek:** Kullanıcıdan bir kenar uzunluğu alarak turtle modülü ile beşgen çizmek için bir fonksiyon yazalım.

````
def besgen_ciz(kenar_uzunlugu):
    """
    Kullanıcıdan alınan kenar uzunluğuna göre turtle modülü ile beşgen çizer.

    Args:
    kenar_uzunlugu (int): Beşgenin bir kenarının uzunluğu
    """
    for _ in range(5): # Beşgen 5 kenera sahip olduğu için 5 defa itere edeceğiz
        kalem.forward(kenar_uzunlugu)
        # Beşgenin iç açısı 108 derece, dış açısı 72 derece
        # left(108)
        kalem.right(72)
````

````
# Kullanıcıdan kenar uzunluğunu al
kenar_uzunlugu = int(input("Beşgenin bir kenar uzunluğunu girin: "))

# Beşgeni çiz
kalem.reset()
besgen_ciz(kenar_uzunlugu)
````

### Örnek: `pencolor`, `pensize` ve `color` Kullanımı

Aşağıdaki örnekte, turtle modülünü kullanarak farklı renk ve kalınlıkta çizimler yapacağız. Bu örnek, Colab ortamında çalışmak için `ColabTurtlePlus` kullanılarak güncellenmiştir.

````
# ColabTurtlePlus'ı başlat
turtle.initializeTurtle()

# Kaplumbağa objesi oluştur ve 'kalem' değişkenine ata
kalem = turtle.Turtle()

# Kalem rengini sarı yapalım
kalem.pencolor("yellow")
# Kalem kalınlığını 3 piksel yapalım
kalem.pensize(3)
# 100 piksel ileri gidelim
kalem.forward(100)

# Kaplumbağayı sola 90 derece döndürelim
kalem.left(90)
# Kalem rengini HTML renk kodu ile maviye ayarlayalım
kalem.color("#0000FF")
# Kalem kalınlığını 8 piksel yapalım
kalem.pensize(8)
# 100 piksel ileri gidelim
kalem.forward(100)

# Kaplumbağayı tekrar sola 90 derece döndürelim
kalem.left(90)
# Kalem rengini sarı yapalım
kalem.pencolor("green")
# Kalem kalınlığını 6 piksel yapalım
kalem.pensize(6)
# 100 piksel ileri gidelim
kalem.forward(100)

# Kaplumbağayı tekrar sola 90 derece döndürelim
kalem.left(90)
# Kalem rengini HTML renk kodu ile ayarlayalım
kalem.color("red")
# Kalem kalınlığını 1 piksel yapalım
kalem.pensize(1)
# 100 piksel ileri gidelim
kalem.forward(100)
````

1. **Kalem Nesnesi Oluşturma**: `kalem = turtle.Turtle()` komutu ile bir turtle nesnesi oluşturuyoruz.
2. **Kalem Rengini Ayarlama**: `kalem.pencolor("red")` ve `kalem.color("#ACE515")` komutları ile kalem rengini değiştiriyoruz.
3. **Kalem Kalınlığını Ayarlama**: `kalem.pensize(3)` ve `kalem.pensize(8)` komutları ile kalem kalınlığını değiştiriyoruz.
4. **Çizim Yapma**: `kalem.forward(100)` ve `kalem.left(90)` komutları ile çizim yapıyoruz.

Bu örnek ile turtle modülünde renk ve kalem kalınlığını ayarlamayı öğrenmiş oldunuz. Farklı renk ve kalınlık değerlerini deneyerek kendi çizimlerinizi oluşturabilirsiniz.

### Turtle Modülünde İleri Seviye Fonksiyonlar

Turtle modülünü kullanarak grafik çizimlerini daha etkili bir şekilde yönetmek için çeşitli fonksiyonlar bulunmaktadır. Bu fonksiyonlar, kaplumbağanın belirli noktalara gitmesini, nokta çizmesini, çizim yapmadan ilerlemesini ve çizime tekrar dönmesini sağlar. Ayrıca, kaplumbağanın görünümünü değiştirmek için de çeşitli seçenekler sunar.

#### 1. Kaplumbağayı Belirli Bir Noktaya Götürmek: `goto(x, y)`

`goto()` fonksiyonu, kaplumbağayı belirtilen (x, y) koordinatlarına taşır. **Bu taşıma süresince çizmeye devam eder.** Çizim yapmadan belli bir koordinata taşımak için önce `.penup()` metodu ile kalemi kaldırırız sonra `goto()`

````
kalem.reset()
kalem.goto(100, 200)  # Kaplumbağa (100, 200) noktasına çizerek gider
````

````
kalem.goto
````

#### 2. Nokta Çizmek: `dot(size, color)`

`dot()` fonksiyonu, belirtilen büyüklükte ve renkte bir nokta çizer.

````
kalem.reset()
kalem.dot(20, "blue")  # 20 piksel büyüklüğünde mavi bir nokta çizer
````

#### 3. Çizmeden İlerlemek: `penup()` veya `up()`

`penup()` veya `up()` fonksiyonu, kaplumbağanın çizim yapmadan hareket etmesini sağlar.

````
kalem.reset()
kalem.penup()  # Kalemi kaldırır, çizim yapmadan ilerler
kalem.forward(100)  # 100 piksel ileri gider
````

#### 4. Çizime Tekrar Başlamak: `pendown()` veya `down()`

`pendown()` veya `down()` fonksiyonu, kaplumbağanın tekrar çizim yapmaya başlamasını sağlar.

````
kalem.reset()
kalem.pendown()  # Kalemi indirir, çizim yaparak ilerler
kalem.forward(100)  # 100 piksel ileri
````

#### 5. Kaplumbağanın Görünümünü Değiştirmek: `shape()`

`shape()` fonksiyonu, kaplumbağanın görünüm şeklini değiştirir. Bu fonksiyon, ‘arrow’, ‘classic’, ‘turtle’ ve ‘circle’ gibi dört şekli destekler.

````
kalem.reset()
kalem.shape("turtle")  # Kaplumbağanın görünümünü 'turtle' yapar
````
