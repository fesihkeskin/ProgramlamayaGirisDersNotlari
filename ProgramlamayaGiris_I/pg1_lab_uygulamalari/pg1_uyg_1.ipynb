**Ödev soru 1:** Kullanıcıdan alfabenin bir harfini girmesini isteyen bir program oluşturacaksınız. Kullanıcı a, e, i, o veya u girerse, programınız girilen harfin sesli olduğunu belirten bir mesaj görüntülemelidir. Kullanıcı y'yi girerse, y'nin sessiz olduğunu belirten bir mesaj görüntülemelidir. Harflerin küçük karakter olarak girildiği varsayılmalıdır (Örn: e, z, a, t gibi)

```python
harf = input("harfi giriniz:")
if harf == ("a" or "e" or "ı" or "i" or "o" or "u" or "ü"):
  print("sesli harftir")
else:
  print("sessi harftir")
```

**Ödev soru 2:** Bir ayın uzunluğu 28 ile 31 gün arasında değişmektedir. Kullanıcıdan bir ayın adını soran bir program oluşturunuz ve o aydaki gün sayısını ekrana yazdırınız. Artık yılların ele alınması için Şubat için "28 veya 29 gün" olarak varsayabilirsiniz.

```python
ay = input("insert month:").lower()

if ay==("ocak" or "mart" or "mayıs" or "temmuz" or "ağustos" or "ekim" or"aralık"):
  print("31 gün çeker")
elif ay==("şubat"):
  print("28 gün çeker")
else:
  print("30 gün çeker")
```

**Ödev soru 3:** Bir üçgen, kenarlarının uzunluklarına göre eşkenar, ikizkenar veya çeşitkenar üçgen olarak sınıflandırılabilir. Bir eşkenar üçgenin 3 kenarının tamamı aynı uzunluktadır. Bir ikizkenar üçgenin aynı uzunlukta iki kenarı ve farklı uzunlukta bir üçüncü kenarı vardır. Tüm kenarların farklı uzunlukları varsa, üçgen çeşitkenardır. Kullanıcıdan bir üçgenin 3 kenarının uzunluğunu girmesini isteyen bir program yazınız ve girilen değerlere göre üçgenin türünü ekranda yazdırınız.

```python
x=int(input("1.kenarı giriniz:"))
y=int(input("2.kenarı giriniz :"))
z=int(input("3.kenarı giriniz :"))
if x==y==z:
  print("eşkenar üçgendir")
elif x == y != z or x == z != y or z == y != x:
  print("ikizkenar üçgendir")
else:
  print("çeşitkenar üçgendir")
```

**Ödev soru 4:** Çin de, hayvanları 12 yıllık bir döngüde yıllara isim vermek için atamaktadırlar. Bir sonraki sayfadaki tabloda 12 yıllık bir döngü gösterilmektedir. Yıl isimleri 12 yılda bir tekrar etmektedir. Yani 2012 bir ejderha yılı ve 1999 da bir tavşan yılı. Kullanıcıdan bir yıl girmesini isteyen ve o yılla ilişkili hayvanı gösteren bir program yazın. Programınız, yalnızca tabloda listelenen yılları değil, sıfırdan büyük veya sıfıra eşit herhangi bir yıl için doğru şekilde çalışmalıdır.

İpucu: Girilen yılın 12’ e bölümünden kalana göre işlem yapınız. Örn:

* 2004 yada 2016 yılı 12 e bölündüğünde kalan 0 (sıfır) ise Maymun yılı,
* 2008 yada 2020 12 e bölündüğünde kalan 4 ise Fare yılı,
* 2010 yada 2022 yılı 12 e bölündüğünde kalan 6 ise Kaplan yılı gibi düşünülebilir.

```python
# Çin zodyak hayvanlarını bir listeye atayalım.
zodyak_hayvanlari = [
    "Maymun",  # 0
    "Horoz",   # 1
    "Köpek",   # 2
    "Domuz",   # 3
    "Fare",    # 4
    "Öküz",    # 5
    "Kaplan",  # 6
    "Tavşan",  # 7
    "Ejderha", # 8
    "Yılan",   # 9
    "At",      # 10
    "Koyun"    # 11
]

# Kullanıcıdan yıl girmesini isteyelim.
yil = int(input("Lütfen bir yıl giriniz: "))

# Girilen yılın 12'ye bölümünden kalanı bulalım.
kalan = yil % 12

# Kalan değerine göre zodyak hayvanını ekrana yazdıralım.
zodyak_hayvani = zodyak_hayvanlari[kalan]
print(f"Girdiğiniz yıl {yil}, bir {zodyak_hayvani} yılıdır.")
```

**Ödev soru 5:** Bir şirkette, çalışanlar her yılın sonunda değerlendirilmektedir. Değerlendirme
ölçeği 0-10 arasında puanlardan oluşmaktadır. Yüksek değerler daha iyi
performansı gösterir ve maaşta daha çok zam ile sonuçlanır. Performansa göre
zam tutarı aşağıdaki tabloda gösterilmiştir.

**puan | zam oranı**

0,1,2   |0

3,4,5   |0.1

6,7,8  |0.2

9,10     |0.3


Bir çalışanın zam miktarı, aldığı puan karşılığındaki zam oranının 2000 TL ile
çarpılmasıyla bulunur.

```python
# Kullanıcıdan çalışanın performans puanını alalım.
puan = int(input("Lütfen çalışanın performans puanını giriniz (0 - 10): "))

# Zam oranını ve miktarını hesaplayalım.
zam_miktari = 0

if 0 <= puan <= 2:
    zam_miktari = 2000 * 0  # Zam yapılmayacak.
elif 3 <= puan <= 5:
    zam_miktari = 2000 * 0.1
elif 6 <= puan <= 8:
    zam_miktari = 2000 * 0.2
elif 9 <= puan <= 10:
    zam_miktari = 2000 * 0.3
else:
    print("Geçersiz puan girdiniz. Lütfen 0 ile 10 arasında bir değer giriniz.")

# Zam miktarını ekrana yazdıralım.
if 0 <= puan <= 10:
    print(f"Çalışanın zam miktarı: {zam_miktari} TL.")
```
