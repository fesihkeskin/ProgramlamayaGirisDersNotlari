#Alıştırma 1:

Yas ve isim parametreleri alan ve aşağıdaki durumlara göre mesajları ekrana yazan karsilama2 () fonksiyonunu yazınız.

Yaşı 10 ve 10 dan küçük olanlara “Merhaba çocuk ‘isim’, Hoş geldin.”

Yaşı 10 ile 30 arasında olanlara “Merhaba genç ‘isim’, Hoş geldin.”

Yaşı 30 ve 30 dan büyük olanlara “Merhaba efendim ‘isim’, Hoş geldin.”

```python
def karsilama2(isim, yas):
    if yas < 10:
        print(f"Merhaba çocuk {isim}, Hoş geldin.")
    elif 10 <= yas <= 30:
        print(f"Merhaba genç {isim}, Hoş geldin.")
    else:
        print(f"Merhaba efendim {isim}, Hoş geldin.")

# Kullanım örneği:
isim = input("İsminizi girin: ")
yas = int(input("Yaşınızı girin: "))
karsilama2(isim, yas)
```

# Alıştırma 2:

sayi isimli tek bir parametre alan ve girilen sayinin faktöriyelini bulup, aşağıdaki ekran çıktısı şeklinde ekrana yazan faktoriyel() fonksiyonunu yazınız.

Not1:1 den n ye kadar olan tamsayıların çarpımına n faktöriyel denir ve n! şeklinde gösterilir.

Örn: 5!=1.2.3.4.5=120

Not2: 0!=1 dir.

```python
def faktoriyel(sayi):
    if sayi == 0:
        return 1
    else:
        sonuc = 1
        for i in range(1, sayi + 1):
            sonuc *= i
        return sonuc

# Kullanım örneği:
sayi = int(input("Faktöriyelini hesaplamak istediğiniz sayıyı girin: "))
sonuc = faktoriyel(sayi)
print(f"{sayi}! = {sonuc}")
```

# Alıştırma 3:

Taksi ücretleri 8TL biniş ücreti artı her 140 metrede 0.35 TL  ücretten oluşur. Tek parametresi olarak gidilen mesafeyi (kilometre cinsinden) alan ve sonuç olarak toplam ücretini veren bir fonksiyon yazınız.

```python
def taksi_ucreti(mesafe_km):
    binis_ucreti = 8  # Binis ucreti
    ucret_her_metre = 0.35 / 140  # Her 140 metrede 0.35 TL ek ucret

    toplam_ucret = binis_ucreti + (mesafe_km * 1000 * ucret_her_metre)

    return toplam_ucret

# Kullanım örneği:
mesafe_km = float(input("Gidilen mesafeyi (kilometre cinsinden) girin: "))
ucret = taksi_ucreti(mesafe_km)
print(f"Toplam taksi ücreti: {ucret} TL")
```

# Alıştırma 4:

Hiçbir parametre almayan ve kullanıcıdan üç sayı girmesini isteyen ve sonuç olarak bu değerlerin medyan değerini veren bir fonksiyon yazınız.

İpucu: Medyan, üç değer küçükten büyüğe doğru sıralandığında ortada olan değerdir.  Örn: 4, 8, 2 değerlerinin medyanı için önce 2, 4, 8 olarak sıralanır ve 4 değeri medyandır.

```python
def medyan_hesapla():
    sayi1 = float(input("Birinci sayıyı girin: "))
    sayi2 = float(input("İkinci sayıyı girin: "))
    sayi3 = float(input("Üçüncü sayıyı girin: "))

    # Sayıları sırala
    sirali_sayilar = sorted([sayi1, sayi2, sayi3])

    # Medyanı hesapla
    medyan = sirali_sayilar[1]  # Orta değer

    return medyan

# Kullanım örneği:
medyan_degeri = medyan_hesapla()
print(f"Girilen sayıların medyanı: {medyan_degeri}")
```

# Alıştırma 5:
Çoğu kişi, özellikle akıllı telefonlar gibi küçük cihazlarda yazı yazarken büyük harfleri doğru kullanmaz. Bu soruda, bir stringdeki uygun karakterleri büyük harfle yazan donuştur adına bir fonksiyon yazınız. Stringdeki ilk karakter büyük harfle yazılmalıdır. Küçük "i" harfi, önünde de boşluk varsa, büyük «I» harfi ile değiştirilmelidir. İpucu: Bir string methodu olan replace inceleyiniz.

```python
def donustur(text):
    """
    Verilen bir metni düzenleyerek, belirli kurallara göre büyük harflerle yazar.

    - Metnin ilk harfi büyük harf olacak.
    - Eğer bir boşluk karakterinden sonra 'i' harfi geliyorsa, bu 'i' harfi büyük 'I' ile değiştirilecek.

    Parametreler:
    text (str): Düzenlenecek metin.

    Döndürülen Değer:
    str: Düzenlenmiş metin.
    """
    # Metnin ilk harfini büyük yap
    text = text.capitalize()

    # Metindeki boşluk+ı kombinasyonlarını büyük I ile değiştir
    text = text.replace(' i', ' I')

    return text
```

```python
# Fonksiyonu test edelim
ornek_metin = "bu bir test metnidir. İşte böyle yazılır."
donustur(ornek_metin)
```

```python
ornek_metin2 = "ne zaman idris orada olur?"
donustur(ornek_metin2)
```

# Alıştırma 6:

Rastgele bir parola oluşturan bir fonksiyon yazınız. Parola 7 ile 10 karakter arasında rastgele uzunlukta olmalıdır. Her karakter, ASCII tablosunda 33 ila 126. konumlardan rastgele seçilmelidir. Fonksiyonunuz herhangi bir parametre almayacaktır. Tek sonucu olarak rastgele oluşturulan parolayı döndürecektir. İpucu: Bu soruyu yaparken `chr( )` fonksiyonu işinize yarayacaktır. Bu fonksiyon ile ilgili ayrıntılı bilgi çevrimiçi olarak mevcuttur.

```python
import random
def rastgele_parola_olustur():
    """
    Rastgele bir parola oluşturur. Parolanın uzunluğu 7 ile 10 karakter arasında rastgele seçilir.
    Her karakter, ASCII tablosunun 33 ila 126 arasındaki karakterlerinden rastgele seçilir.

    Parametreler:
    Yok

    Döndürülen Değer:
    str: Rastgele oluşturulan parola.
    """
    parola_uzunlugu = random.randint(7, 10)
    parola = ""

    for _ in range(parola_uzunlugu):
        karakter = chr(random.randint(33, 126))
        parola += karakter

    return parola
```

```python
# Fonksiyonu test edelim
rastgele_parola_olustur()
```

Bu fonksiyon, her çağrıldığında 7 ile 10 karakter uzunluğunda, ASCII tablosunun 33 ila 126 arasındaki karakterlerden rastgele seçilen karakterlerle bir parola oluşturur. Her çağrıda farklı bir parola üretir

# Alıştırma 7:

Bir parolanın güçlü olup olmadığını belirleyen bir fonksiyon yazınız. Güçlü parola; en az 8 karakter uzunluğunda, en az bir büyük ve bir küçük harf ve, en az bir rakam içeren bir paroladır. Fonksiyon, tek parametre olarak kendisine iletilen parolayı alır ve güçlü bir parola ise «Güçlü parola» değilse «Zayıf parola» ifadelerini ekrana yazar.

```python
# alternatif çözüm:
def parola_guclu_mu1(parola):
    """
    Verilen parolanın güçlü olup olmadığını kontrol eder.

    Güçlü bir parola için şartlar:
    - En az 8 karakter uzunluğunda
    - En az bir büyük harf içermeli
    - En az bir küçük harf içermeli
    - En az bir rakam içermeli

    Parametreler:
    parola (str): Kontrol edilecek parola.

    Döndürülen Değer:
    str: Parolanın güçlü olup olmadığına dair mesaj.
    """
    if len(parola) < 8:
        return "Zayıf parola - Uzunluk yetersiz"

    if not any(char.isupper() for char in parola):
        return "Zayıf parola - Büyük harf eksik"

    if not any(char.islower() for char in parola):
        return "Zayıf parola - Küçük harf eksik"

    if not any(char.isdigit() for char in parola):
        return "Zayıf parola - Rakam eksik"

    return "Güçlü parola"
```

```python
# Fonksiyonu test edelim
ornek_parola = "Python123"
parola_guclu_mu1(ornek_parola)
```

```python
def parola_guclu_mu2(parola):
    """
    Verilen parolanın güçlü olup olmadığını kontrol eder.

    Güçlü bir parola şu özelliklere sahip olmalıdır:
    - En az 8 karakter uzunluğunda olmalı.
    - En az bir büyük harf içermeli.
    - En az bir küçük harf içermeli.
    - En az bir rakam içermeli.

    Parametreler:
    parola (str): Kontrol edilecek parola.

    Döndürülen Değer:
    str: Parolanın güçlü olup olmadığını belirten ifade.
    """
    buyuk_harf = any(c.isupper() for c in parola)
    kucuk_harf = any(c.islower() for c in parola)
    rakam = any(c.isdigit() for c in parola)

    if len(parola) >= 8 and buyuk_harf and kucuk_harf and rakam:
        return "Güçlü parola"
    else:
        return "Zayıf parola"
```

```python
# Fonksiyonu test edelim
ornek_parolalar = ["Guv3nli123", "parola", "ParolA123", "1234567"]
for p in ornek_parolalar:
    sonuc = parola_guclu_mu2(p)
    print(f"'{p}': {sonuc}")
```

`parola_guclulugu_kontrol` fonksiyonunun verdiği örnek sonuçlar:

- 'Guv3nli123': Güçlü parola
- 'parola': Zayıf parola
- 'ParolA123': Güçlü parola
- '1234567': Zayıf parola

Bu fonksiyon, bir parolanın en az 8 karakter uzunluğunda olup olmadığını, en az bir büyük ve bir küçük harf ile en az bir rakam içerip içermediğini kontrol ederek, parolanın güçlü olup olmadığını belirler.
