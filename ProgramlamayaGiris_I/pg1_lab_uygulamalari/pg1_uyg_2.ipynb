**Soru 1:** 1’den 100'e kadar olan sayıları yanyana yazdırın:


````python
sayi = 1  # Başlangıç değeri olarak 1'i atadık.
while sayi <= 100:  # sayi değişkeni 100 veya daha küçük olduğu sürece döngü devam eder.
    print(sayi, end=' ')  # print fonksiyonunun end parametresi sayıların arasına boşluk koymamızı sağlar.
    sayi += 1  # sayi değişkenini her adımda 1 artırıyoruz.
````

Bu sayede tüm sayılar aynı satır üzerinde yanyana yazdırılır.

**Soru 4:** Kullanıcıya “Türkiye’nin başkenti hangi şehirdir?” sorusunu sorsun. Doğru cevap verilene kadar “Yanlış cevap, tekrar deneyiniz!” uyarısını versin, Doğru cevap verildiğinde “Tebrikler, bildiniz” yazısı gösterilsin ve döngüden çıkılsın.

````python
while True:  # Sonsuz döngü
    cevap = input("Türkiye'nin başkenti hangi şehirdir? ").strip().lower()
    # Kullanıcıdan cevap alınır,.strip() ile boşluklar silinir ve .lower() ile küçük harfe çevrilir
    if cevap == "ankara":  # Doğru cevap kontrolü
        print("Tebrikler, bildiniz!")
        break  # Doğru cevap verildiğinde döngüden çıkılır
    else:
        print("Yanlış cevap, tekrar deneyiniz!")  # Yanlış cevap verildiğinde uyarı mesajı verilir
````

Bu kod, kullanıcı doğru cevabı verene kadar tekrar edilir. Doğru cevap (`"ankara"`) girildiğinde, "Tebrikler, bildiniz!" mesajını yazdırır ve `break` deyimi ile döngüyü sonlandırır. Yanlış cevap verildiğinde ise kullanıcıyı tekrar denemesi için uyarır. Kullanıcıdan alınan cevap `strip()` metodu ile başındaki ve sonundaki boşluklardan arındırılır ve `lower()` metodu ile tümü küçük harfe çevrilir, böylece kullanıcının girdiği cevabın büyük/küçük harf kullanımından etkilenmemesini sağlarız.

**Soru 5:** Faktöriyel: dışardan girilen bir sayının faktöriyelini hesaplayınız.

````python
# Kullanıcıdan bir sayı alıyoruz
sayi = int(input("Faktöriyelini hesaplamak istediğiniz sayıyı girin: "))

# Faktöriyel hesaplamak için başlangıç değerleri belirliyoruz
faktoriyel = 1
i = 1

# sayi'nin faktöriyelini hesaplıyoruz
while i <= sayi:
    faktoriyel *= i
    i += 1

# Sonucu ekrana yazdırıyoruz
print(f"{sayi} sayısının faktöriyeli: {faktoriyel}")
````

Bu programda, kullanıcıdan bir sayı istenir ve `faktoriyel` değişkeni 1 olarak başlatılır. `while` döngüsü, `i` değişkeni 1'den başlayarak kullanıcının girdiği sayıya eşit veya daha büyük olana kadar her adımda `i` değeri ile `faktoriyel` değişkenini çarpar ve `i`'yi 1 artırır. Döngü tamamlandığında, `faktoriyel` değişkeni, girilen sayının faktöriyel değerini içerir ve bu değer ekrana yazdırılır.

**Soru 6:** Adını daha sonra soyadını doğru yazana kadar While döngüsünü kullanın.

````python
dogru_ad = "Ahmet"
dogru_soyad = "Yılmaz"

# Ad doğru girilene kadar döngü
while True:
    ad = input("Adınızı girin: ")
    if ad == dogru_ad:
        print("Doğru ad!")
        break
    else:
        print("Yanlış ad, tekrar deneyiniz!")

# Soyad doğru girilene kadar döngü
while True:
    soyad = input("Soyadınızı girin: ")
    if soyad == dogru_soyad:
        print("Doğru soyad!")
        break
    else:
        print("Yanlış soyad, tekrar deneyiniz!")
print("Kullanıcı doğru.")
````

Hemen üstteki kodda, kullanıcı doğru adı girinceye kadar ilk `while` döngüsü devam eder. Doğru ad girildiğinde `break` ifadesi ile ilk döngüden çıkılır ve soyadı sormak için ikinci döngü başlar. Soyad da doğru girildiğinde ikinci döngü de `break` ile sonlanır ve program sona erer. Bu süre boyunca kullanıcı yanlış bir giriş yaptığında, döngü içindeki uygun uyarı mesajı ekrana yazdırılır ve doğru bilgi girilene kadar döngü devam eder.

**Soru 7:** Ekrana 10 tane rastgele sayı yazdırın.

````python
import random

sayac = 0  # Sayacı başlat

while sayac < 10:
    rastgele_sayi = random.randint(1, 100)  # 1 ile 100 arasında rastgele bir sayı üret
    print(rastgele_sayi)
    sayac += 1  # Her adımda sayacı 1 artır
````

Python'da rastgele sayı üretmek için `random` modülünü kullanabiliriz. `while` döngüsü ile 10 tane rastgele sayı yazdırmak için şu adımları takip edebiliriz:

1. `random` modülünü içe aktarın.
2. Bir sayaç tanımlayın.
3. `while` döngüsü içinde, sayaç 10'a ulaşana kadar rastgele sayılar üretin ve ekrana yazdırın.


Bu kod parçası, 1 ile 100 arasında (her iki sınır da dahil) 10 adet rastgele sayı üretir ve her bir sayıyı ekrana yazdırır. `random.randint(a, b)` fonksiyonu, `a` ve `b` arasında (sınırlar dahil) rastgele bir tamsayı döndürür. `sayac` her iterasyonda 1 artırılarak döngünün 10 kez çalışmasını sağlar.

**Soru 8:** Ekrana 10 tane rastgele tek sayı yazdırın.

````python
import random # random modülünü daha önce içe aktardıysak buraya tekrar yazmamıza gerek yok

sayac = 0  # Sayacı başlat

while sayac < 10:
    rastgele_sayi = random.randint(1, 100)  # 1 ile 100 arasında rastgele bir sayı üret
    if rastgele_sayi % 2 == 1:  # (ya da rastgele_sayi %2 != 0): Sayının tek olduğunu kontrol et.
        print(rastgele_sayi)
        sayac += 1  # Tek sayı üretildiğinde sayacı 1 artır
````

**Soru 9:** Ekrana 1 tek 1 çift olmak üzere 10 tane rasgele sayı yazınız.

````python
import random

sayac = 0

while sayac < 10:
    rastgele_sayi = random.randint(1, 100)
    if rastgele_sayi % 2 != 0:  # Tek sayı kontrolü
        print(rastgele_sayi)
        sayac += 1  # Tek sayı yazdırıldıktan sonra sayacı artır
    else:
        continue  # Eğer çift sayı ise, döngünün başına dön

    rastgele_sayi = random.randint(1, 100)
    while True:
        if rastgele_sayi % 2 == 0:  # Çift sayı kontrolü
            print(rastgele_sayi)
            sayac += 1  # Çift sayı yazdırıldıktan sonra sayacı artır
            break  # Çift sayı bulunduğunda while döngüsünden çık
        else:
            rastgele_sayi = random.randint(1, 100)  # Eğer tek sayı ise, yeni bir rastgele sayı üret
````

Bu kod, her seferinde önce bir tek sayıyı kontrol eder ve eğer sayı tekse ekrana yazdırır ve sayacı artırır. Daha sonra bir çift sayı bulmak için yeni bir rastgele sayı üretilir. Eğer bu sayı çiftse, o da ekrana yazdırılır ve sayacı artırılır. Eğer sayı tekse, çift sayı bulunana kadar yeni rastgele sayılar üretilmeye devam eder. Bu işlem, sayacın değeri 10'a ulaşana kadar devam eder.

**Soru 11:**

Asal sayılar, yalnızca 1 ve kendisi ile tam olarak bölünebilen pozitif tam sayılardır. Şimdi sizden, kullanıcının girdiği bir sayının asal olup olmadığını kontrol eden ve sonucu ekrana yazdıran bir Python programı yazmanızı istiyoruz.

**NOT:** Yazacağınız programı sadece işlediğimiz konuları kullanarak yazmanızı istiyoruz. Yani sorunun çözümününde `if`, `elif`, `else` ve `while` gibi sadece işlediğimiz konulardan faydalanacaksınız. Henüz işlemediğimiz, fonksiyonlar (`def`) ve `for` döngüleri gibi, yöntemlerin kullanımı kabul edilmeyecektir. Bu şartı sağlamadığınız durumda bu ödevden **sıfır** alırsınız.

**Beklenen Çıktı:**

Eğer girilen sayı asal sayı **değil** ise beklenen örnek ekran çıktısı:

```
Bir sayı giriniz: 4
4 sayısı asal değildir!
4 sayısının bölenleri şunlardır: [1, 2, 4]
```

Eğer girilen sayı asal sayı ise beklenen örnek ekran çıktısı:

```
Bir sayı giriniz: 5
5 sayısı asaldır!
```

**İpucu:**
Bu sorunun çözümü için farklı yöntemler olmakla birlikte aşağıdaki adımları izleyerek kolayca çözebilirsiniz:

1. **Asal sayı tanımı üzerine düşünün:** Asal sayılar yalnızca 1 ve kendisi (girilen `sayi`) ile bölünebilen sayılardır. Asal sayının tanımından hareketle bir sayının asallığını nasıl kontrol edebileceğinizi düşünün.

2. **Bölen sayısını kontrol edin:** Bir sayının asal olup olmadığını anlamak için, `bolen = 1`'den başlayarak o sayıya kadar olan tüm sayılarla bölünüp bölünemediğini kontrol edin (yani `bolen` arttırılacak). Eğer sayı, yalnızca 1 ve kendisi dışında başka bir sayıya bölünüyorsa (yani kaç sayıya bölündüğünü saymanız gerekiyor `sayac = 0`'dan başlatıp artırırız), o sayı asal değildir (yani `sayac>2` ise asal değildir).

3. **Döngü kullanın:** Sayının tüm bölenlerini bulmak için `while` döngüsünü kullanın. Döngü ile, `bolen = 1`'den başlayarak verilen sayıya kadar (girilen `sayi`  dahil) tüm sayıları kontrol etmelisiniz.

4. **Mod operatörü:** Mod operatörü (`%`), bir sayının başka bir sayıya bölümünden kalanı verir (`kalan = sayi % bolen`). Sayının her bir böleni ile bölünüp bölünmediğini kontrol etmek için mod operatörünü kullanabilirsiniz.

5. **Bölen sayısını sayın:** Asal olmayan bir sayının en az üç pozitif böleni olacağını (1, kendisi ve en az bir tane daha) unutmayın. Eğer bölenlerin sayısı 2'den fazla ise, sayının asal olmadığını belirtebilirsiniz.

6. **Bölenleri bir yere yazın:** Eğer sayı asal değilse girilen sayının tüm pozitif bölenlerini yazacağımız `bolenler = []` gibi boş bir listeye ihtiyacımız olacak. Bu listeye bölenleri ekleyeceğiz (`bolenler.append(bolen)`) ve bu listeyi kullanarak bu bölenleri ekrana yazdırabilirsiniz.



````python
sayi = int(input("Bir sayı giriniz: "))
bolen = 1 # Asallık kontrolü için 1'den başladık.
sayac = 0 # Kaç tane pozitif bölen olduğunu saymak için sayaç.
bolenler = [] # Bölenleri kaydedeceğimiz boş bir liste oluşturduk.
while bolen <= sayi:  # Girilen sayıya kadar tüm bölenleri deneme döngüsü
  kalan = sayi % bolen  # bölen'e tam bölünüp bölünmediği kontrolü için kalanı bulma
  if kalan == 0:  # eğer kalan sıfırsa bölen'e tam bölünüyor
    sayac += 1  # if'in içinde sayacı bir arttır
    bolenler.append(bolen)  #if'in içinde böleni listeye ekle
  bolen += 1  # if'in dışına çıkıldı, burası while döngüsünün içi
              # bir sonraki böleni denemek üzere bir arttırıp while'a geri dön

# aşağıdaki kodlar while döngüsünün dışıdır
if sayac > 2 or sayi == 1:  # Asal sayılar ve 1 için kontrol
  print("%d sayısı asal değildir!" %(sayi))
  print("%d sayısının bölenleri şunlardır: %s"%(sayi,bolenler))
else:
  print("%d sayısı asaldır!" %(sayi))
````
