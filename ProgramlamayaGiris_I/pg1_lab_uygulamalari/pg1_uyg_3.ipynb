**Soru 1:** Bir sayı dizisinin ortalamasını hesaplayan bir programımız var. Sizden bir sayı girmenizi ve ardından isterseniz başka bir sayı daha girmenizi isteyeceğiz. Sayı girişlerinizi bitirmek ve ortalamayı hesaplamak istediğinizde, programın size soracağı "Çıkmak ister misiniz?" sorusuna "Evet" yanıtını vermeniz yeterli. Eğer devam etmek istiyorsanız "Hayır" demeniz ve ardından yeni sayınızı girmeniz beklenir.

````python
toplam = 0
sayac = 0

while True:
    sayi = float(input("Lütfen bir sayı giriniz: "))
    toplam += sayi
    sayac += 1

    cikis = input("Çıkmak ister misiniz? (Evet/Hayır): ").strip().lower()
    if cikis == 'evet':
        break

# Eğer sayı girilmişse ortalamayı hesapla ve yazdır
if sayac > 0:
    ortalama = toplam / sayac
    print(f"Girdiğiniz sayıların ortalaması: {ortalama:.2f}")
else:
    print("Hiçbir sayı girilmedi.")
````

**Soru 2:**
Bir hayvanat bahçesinin fiyatlandırma politikası, ziyaretçilerin yaş gruplarına göre değişiklik göstermektedir ve sizden, bir Python programı yazarak bu politikanın uygulanmasını otomatize etmenizi istiyoruz.

İşte kurallar:
- 2 yaş ve altı için giriş ücretsiz,
- 3 ile 12 yaş arası için 14 TL,
- 65 yaş ve üstü için 18 TL,
- Diğer tüm yaş grupları için ise 23 TL giriş ücreti belirlenmiştir.

Sizin göreviniz, bir gruptaki kişilerin yaşlarını girmelerini sağlayacak ve bu verilere dayanarak toplam giriş ücretini hesaplayacak bir program geliştirmektir. Programınız, kullanıcıdan grupta kaç kişi olduğunu almalı ve ardından her bir bireyin yaşını sırayla sormalıdır. Kullanıcı gruptaki tüm kişilerin yaşını girdikten sonra programınız grubun toplam giriş ücretini hesaplayarak ekrana yazdırsın.



````python
# Toplam ücreti saklamak için bir değişken oluştur.
toplam_ücret = 0

# Kullanıcıdan kişi sayısını al.
kisi_sayisi = int(input("Lütfen kaç kişi olduğunuzu giriniz: "))

# Her bir grup üyesinin yaşını alıp ücreti hesapla.
for i in range(kisi_sayisi):
    yas = int(input(f"{i+1}. konuğun yaşını giriniz: "))

    # Yaşa göre ücreti hesapla ve toplam ücrete ekle.
    if yas <= 2:
        print("2 yaş ve altı için giriş ücretsizdir.")
    elif 3 <= yas <= 12:
        toplam_ücret += 14
    elif 65 <= yas:
        toplam_ücret += 18
    else:
        toplam_ücret += 23

# Toplam ücreti göster.
print(f"Grubunuzun toplam giriş ücreti: {toplam_ücret} TL'dir.")
````

**Soru 3:**Kullanıcı tarafından klavyeden girilen bir sayının asal sayı olup olmadığını kontrol edip, ekrana yazan Python programını yazınız.

• *Not1:* Yalnız 1 (bir) ve kendisi ile bölünebilen birden büyük doğal sayılar asal sayıdır.

• *Not2:* İlk asal sayı 2 dir.


````python
# Kullanıcıdan bir sayı alın ve asal olup olmadığını kontrol edin.
sayi = int(input("Lütfen bir sayı giriniz: "))
asal_mi = True  # Sayının asal olduğunu varsayarak başlayalım.

# Asal sayı kontrolü yapın.
if sayi < 2:  # Eğer sayı 2'den küçükse, asal sayı olamaz.
    asal_mi = False
else:
    for i in range(2, sayi):  # 2'den başlayarak sayıya kadar olan tüm sayılarla bölünmeyi dene.
        if sayi % i == 0:  # Eğer sayı herhangi bir sayıya tam bölünüyorsa, asal değildir.
            asal_mi = False
            break

# Sonucu yazdırın.
if asal_mi:
    print(f"{sayi} bir asal sayıdır.")
else:
    print(f"{sayi} bir asal sayı değildir.")
````

 `while` döngüsünü kullanarak bir sayının asal olup olmadığını kontrol eden Python programı:

```python
# Kullanıcıdan bir sayı alın ve asal olup olmadığını kontrol edin.
sayi = int(input("Lütfen bir sayı giriniz: "))
asal_mi = True  # Sayının asal olduğunu varsayarak başlayalım.
# Asal sayı kontrolü yapın.
if sayi < 2:  # Eğer sayı 2'den küçükse, asal sayı olamaz.
    asal_mi = False
else:
    i = 2  # Bölme işlemi için sayaç.
    while i < sayi:  # Sayıya kadar olan tüm sayılarla kontrol et.
        if sayi % i == 0:  # Eğer sayı, i'ye tam bölünüyorsa, asal değildir.
            asal_mi = False
            break
        i += 1  # Sayaç değerini artır.
# Sonucu yazdırın.
if asal_mi:
    print(f"{sayi} bir asal sayıdır.")
else:
    print(f"{sayi} bir asal sayı değildir.")
```

Bu programda, kullanıcının girdiği sayının asal olup olmadığını kontrol etmek için `while` döngüsü kullanıyoruz. `if` koşulu ile sayının 2'den küçük olup olmadığına bakıyoruz. Eğer 2'den küçükse, doğrudan `asal_mi` değişkenini `False` olarak ayarlıyoruz ve asal olmadığını söylüyoruz. Eğer sayı 2 veya daha büyükse, `while` döngüsü ile 2'den başlayarak sayının kendisine kadar olan tüm sayılarla bölünebilirliğini kontrol ediyoruz. Eğer sayı, bu aralıktaki herhangi bir sayıya tam bölünüyorsa, yani modu 0 ise, `asal_mi` değişkenini `False` olarak güncelliyoruz ve döngüyü `break` ile sonlandırıyoruz. Bu kontrol tamamlandığında, eğer `asal_mi` hala `True` ise, sayının asal olduğunu söylüyoruz.

````python
# Kullanıcıdan bir kelime alıyoruz.
kelime = input("Lütfen bir kelime giriniz: ").lower()  # Küçük harfe çevirerek karşılaştırma yapılabilir hale getiriyoruz.

# Orijinal kelime ile ters çevrilmiş kelimeyi karşılaştırıyoruz.
if kelime == kelime[::-1]:
    print(f"Girdiğiniz kelime bir palindromdur: {kelime[::-1]}")
else:
    print(f"Girdiğiniz kelime bir palindrom değildir: {kelime[::-1]}")
````

**Soru 5:**

Aritmetik dizilerin temel bir konseptini inceleyeceğimiz bir uygulama yapacaksınız. Aritmetik dizi, her terimin bir öncekine sabit bir fark eklenerek oluşturulduğu bir sayı dizisidir. Örneğin, 1'den başlayarak her seferinde 1 artırarak devam eden bir dizi aritmetiktir.

Bu sorudaki göreviniz, 1'den başlayıp klavyeden girilen bir sayıya kadar olan tüm tam sayıların toplamını hesaplayacak bir Python programı yazmak olacak. Örneğin, eğer 6 girilirse, programınız 1+2+3+4+5+6'nın toplamını hesaplayıp ekrana yazdıracaktır.

````python
# Kullanıcıdan bir sayı alıyoruz.
bitis_sayisi = int(input("Lütfen bir sayı giriniz: "))
toplam = 0  # Toplamı saklamak için bir değişken oluşturuyoruz.

# 1'den başlayarak kullanıcının girdiği sayıya kadar olan sayıları topluyoruz.
for sayi in range(1, bitis_sayisi + 1):
    toplam += sayi

# Sonucu ekrana yazdırıyoruz.
print(f"1'den {bitis_sayisi}'e kadar olan tam sayıların toplamı: {toplam}")
````

Bu program, kullanıcının girdiği sayıya kadar olan tüm tam sayıları toplayarak sonucu ekrana yazdırır. `range` fonksiyonu, 1'den başlayıp kullanıcının girdiği sayıya kadar (bu sayı dahil) olan tüm sayılar için bir dizi oluşturur. Daha sonra, bu dizi üzerinde döngü kullanılarak her bir sayı toplama eklenir. İşlem tamamlandığında, hesaplanan toplam ekrana yazdırılır.
